```python
import os
import numpy as np
import librosa
from sklearn.mixture import GaussianMixture
from sklearn.metrics import roc_curve
import matplotlib.pyplot as plt
from tqdm import tqdm
```

```python
# ---- Utility functions (you likely already have them) ----
def load_data(root, max_clips_per_speaker=None):
    """Return dict: {speaker_name: [feature_matrix,...]}"""
    data = {}
    for spk in sorted(os.listdir(root)):
        spk_path = os.path.join(root, spk)
        if not os.path.isdir(spk_path):
            continue
        feats_list = []
        for i, wav in enumerate(sorted(os.listdir(spk_path))):
            if max_clips_per_speaker and i >= max_clips_per_speaker:
                break
            fpath = os.path.join(spk_path, wav)
            # Replace this with your actual MFCC+LPCC extraction
            mfcc = extract_features(fpath)  # (frames, dims)
            feats_list.append(mfcc)
        if feats_list:
            data[spk] = np.vstack(feats_list)
    return data
```

```python

def train_gmms(features, n_mix):
    gmms = {}
    for sp, feats in features.items():
        n_frames = feats.shape[0]
        if n_frames < n_mix * 10:
            print(f"  [skip {sp}] not enough data ({n_frames} frames)")
            continue
        try:
            gmm = GaussianMixture(
                n_components=n_mix,
                covariance_type='diag',
                reg_covar=1e-3,  # ðŸ”§ increased to stabilize small datasets
                max_iter=300,
                random_state=0
            )
            gmm.fit(feats)
            gmms[sp] = gmm
        except Exception as e:
            print(f"  [skip {sp}] GMM failed: {e}")
    return gmms


def evaluate_eer(gmms, features):
    """Dummy EER-like score: lower = better"""
    speakers = list(gmms.keys())
    scores, labels = [], []
    for sp, feats in features.items():
        if sp not in gmms:
            continue
        for sp2, gmm in gmms.items():
            score = np.mean(gmm.score(feats))
            scores.append(score)
            labels.append(1 if sp == sp2 else 0)
    # Convert to EER-like metric (not exact)
    same = np.mean([s for s, l in zip(scores, labels) if l == 1])
    diff = np.mean([s for s, l in zip(scores, labels) if l == 0])
    eer = max(0, 1 - (same - diff) / abs(same))
    return eer
```

```python
def experiment(data_root):
    # Load full dataset
    full_data = load_data(data_root)
    print(f"Loaded {len(full_data)} speakers.")

    # (i) Unbalanced dataset test
    print("\n(i) Testing effect of unbalanced data:")
    min_len = min([len(v) for v in full_data.values()])
    balanced_data = {k: v[:min_len] for k, v in full_data.items()}
    unbalanced_data = full_data

    gmms_bal = train_gmms(balanced_data, 8)
    gmms_unb = train_gmms(unbalanced_data, 8)

    eer_bal = evaluate_eer(gmms_bal, balanced_data)
    eer_unb = evaluate_eer(gmms_unb, unbalanced_data)

    print(f"Balanced EER = {eer_bal*100:.2f}% | Unbalanced EER = {eer_unb*100:.2f}%")

    # (ii) Effect of number of mixture components
    print("\n(ii) Effect of number of mixture components:")
    features = full_data
    n_components_list = [2, 4, 8, 16]
    eers = []
    for n_mix in n_components_list:
        gmms = train_gmms(features, n_mix)
        eer = evaluate_eer(gmms, features)
        eers.append(eer)
        print(f"  n_mix={n_mix}, EER={eer*100:.2f}%")

    plt.figure()
    plt.plot(n_components_list, [e*100 for e in eers], marker='o')
    plt.title("Effect of Number of Mixture Components")
    plt.xlabel("No. of Mixtures")
    plt.ylabel("EER (%)")
    plt.grid(True)
    plt.show()

    # (iii) Amount of data vs mixture components
    print("\n(iii) Amount of data vs number of components:")
    clip_counts = [2, 4, 8, 12]
    eers_data = {c: [] for c in clip_counts}

    for clips in clip_counts:
        subset_data = load_data(data_root, max_clips_per_speaker=clips)
        for n in n_components_list:
            gmms = train_gmms(subset_data, n)
            eer = evaluate_eer(gmms, subset_data)
            eers_data[clips].append(eer)
            print(f"clips={clips}, components={n}, EER={eer*100:.2f}%")

    # Plot amount of data vs mixture components
    plt.figure()
    for clips, vals in eers_data.items():
        plt.plot(n_components_list, [v*100 for v in vals], marker='o', label=f"{clips} clips")
    plt.title("Amount of Data vs Mixture Components")
    plt.xlabel("Number of Components")
    plt.ylabel("EER (%)")
    plt.legend()
    plt.grid(True)
    plt.show()
```

```python
import os
import numpy as np
import librosa
import matplotlib.pyplot as plt
from sklearn.mixture import GaussianMixture
from sklearn.manifold import TSNE
from sklearn.metrics import roc_curve
from tqdm import tqdm
import random

# ================================
# 1. Data Loading
# ================================
def load_data(data_root, max_clips_per_speaker=None):
    data = {}
    for spk in sorted(os.listdir(data_root)):
        spk_path = os.path.join(data_root, spk)
        if not os.path.isdir(spk_path):
            continue
        files = [f for f in os.listdir(spk_path) if f.endswith('.wav')]
        if max_clips_per_speaker:
            files = files[:max_clips_per_speaker]
        feats = []
        for f in files:
            path = os.path.join(spk_path, f)
            y, sr = librosa.load(path, sr=None)
            mfcc = librosa.feature.mfcc(y=y, sr=sr, n_mfcc=20)
            feats.append(mfcc.T)
        if feats:
            data[spk] = np.vstack(feats)
    return data

# ================================
# 2. Train GMMs
# ================================
def train_gmms(features, n_mix):
    gmms = {}
    for spk, feats in features.items():
        n_frames = feats.shape[0]

        # Adjust mixture count based on available data
        if n_frames < n_mix * 10:
            n_comp = max(2, n_frames // 10)
            print(f"[warn] {spk}: not enough data ({n_frames} frames) â†’ using {n_comp} components instead of {n_mix}")
        else:
            n_comp = n_mix

        try:
            gmm = GaussianMixture(
                n_components=n_comp,
                covariance_type='diag',
                reg_covar=1e-2,        # â† increased for stability
                max_iter=300,
                random_state=0
            )
            gmm.fit(feats)
            gmms[spk] = gmm
        except ValueError as e:
            print(f"[skip] {spk}: GMM fitting failed ({e})")
            continue
    return gmms


# ================================
# 3. Evaluate EER
# ================================
def evaluate_eer(gmms, features):
    scores_true, scores_false = [], []
    speakers = list(gmms.keys())
    for spk in speakers:
        gmm_true = gmms[spk]
        feats_true = features[spk]
        scores_true.extend(gmm_true.score_samples(feats_true))
        for other_spk in speakers:
            if other_spk == spk:
                continue
            feats_other = features[other_spk]
            scores_false.extend(gmm_true.score_samples(feats_other))
    y_true = np.array([1]*len(scores_true) + [0]*len(scores_false))
    y_score = np.array(scores_true + scores_false)
    fpr, tpr, _ = roc_curve(y_true, y_score)
    fnr = 1 - tpr
    eer = fpr[np.nanargmin(np.abs(fnr - fpr))]
    return eer

# ================================
# 4. t-SNE Visualization
# ================================
def visualize_tsne(features):
    print("\nGenerating t-SNE visualization...")
    X, y, labels = [], [], []
    for i, (spk, feats) in enumerate(features.items()):
        idx = np.random.choice(feats.shape[0], size=min(300, feats.shape[0]), replace=False)
        X.append(feats[idx])
        y.extend([i]*len(idx))
        labels.append(spk)
    X = np.vstack(X)
    y = np.array(y)
    tsne = TSNE(n_components=2, perplexity=30, random_state=42)
    X_emb = tsne.fit_transform(X)
    plt.figure(figsize=(8, 6))
    for i, spk in enumerate(labels):
        plt.scatter(X_emb[y == i, 0], X_emb[y == i, 1], label=spk, alpha=0.7)
    plt.title("t-SNE visualization of MFCC features by speaker")
    plt.legend()
    plt.grid(True)
    plt.show()

# ================================
# 5. Main Experiment
# ================================
def experiment(data_root):
    # Load full dataset
    full_data = load_data(data_root)
    print(f"\nLoaded {len(full_data)} speakers.")

    # (i) Effect of Unbalanced Dataset
    print("\n(i) Testing effect of unbalanced dataset:")
    min_len = min(len(v) for v in full_data.values())
    balanced_data = {k: v[:min_len] for k, v in full_data.items()}
    unbalanced_data = full_data

    gmms_bal = train_gmms(balanced_data, 8)
    gmms_unb = train_gmms(unbalanced_data, 8)

    eer_bal = evaluate_eer(gmms_bal, balanced_data)
    eer_unb = evaluate_eer(gmms_unb, unbalanced_data)
    print(f"Balanced EER = {eer_bal*100:.2f}% | Unbalanced EER = {eer_unb*100:.2f}%")

    # (ii) Effect of Number of Mixture Components
    print("\n(ii) Effect of number of mixture components:")
    n_components_list = [2, 4, 8, 16]
    eers_mix = []
    for n_mix in n_components_list:
        gmms = train_gmms(full_data, n_mix)
        eer = evaluate_eer(gmms, full_data)
        eers_mix.append(eer)
        print(f"n_mix={n_mix}, EER={eer*100:.2f}%")

    plt.figure()
    plt.plot(n_components_list, [e*100 for e in eers_mix], marker='o')
    plt.title("Effect of Number of Mixture Components")
    plt.xlabel("Number of Components")
    plt.ylabel("EER (%)")
    plt.grid(True)
    plt.show()

    # (iii) Amount of Data vs Number of Components
    print("\n(iii) Amount of data vs number of mixture components:")
    clip_counts = [2, 4, 8, 12]
    eers_data = {c: [] for c in clip_counts}
    for clips in clip_counts:
        subset_data = load_data(data_root, max_clips_per_speaker=clips)
        for n in n_components_list:
            gmms = train_gmms(subset_data, n)
            eer = evaluate_eer(gmms, subset_data)
            eers_data[clips].append(eer)
            print(f"clips={clips}, components={n}, EER={eer*100:.2f}%")

    plt.figure()
    for clips, vals in eers_data.items():
        plt.plot(n_components_list, [v*100 for v in vals], marker='o', label=f"{clips} clips")
    plt.title("Amount of Data vs Mixture Components")
    plt.xlabel("Number of Components")
    plt.ylabel("EER (%)")
    plt.legend()
    plt.grid(True)
    plt.show()

    # (iv) t-SNE Visualization
    visualize_tsne(full_data)

    print("\nâœ… Experiment completed successfully!")

# ================================
# Entry Point
# ================================
if __name__ == "__main__":
    data_root = "data/data_clips"  # your folder with speaker subfolders
    experiment(data_root)
```

```python
import os
import numpy as np
import librosa
import matplotlib.pyplot as plt
from sklearn.mixture import GaussianMixture
from sklearn.metrics import roc_curve, confusion_matrix, ConfusionMatrixDisplay
from tqdm import tqdm
import argparse
import logging
import warnings

# Suppress warnings from librosa
warnings.filterwarnings('ignore', category=UserWarning, module='librosa')

# ================================
# Setup Logging
# ================================
logging.basicConfig(level=logging.INFO,
                    format='%(asctime)s - %(levelname)s - %(message)s')

# ================================
# 1. Advanced Data Loading & Feature Extraction
# ================================
def extract_features(file_path: str, n_mfcc: int = 20, use_deltas: bool = True, use_vad: bool = True) -> np.ndarray:
    """
    Extracts advanced features from a single audio file.

    Args:
        file_path (str): Path to the audio file.
        n_mfcc (int): Number of MFCCs to compute.
        use_deltas (bool): Whether to append delta and delta-delta features.
        use_vad (bool): Whether to apply Voice Activity Detection.

    Returns:
        np.ndarray: The extracted features, or an empty array if processing fails.
    """
    try:
        y, sr = librosa.load(file_path, sr=None)

        # 1. Voice Activity Detection (VAD)
        if use_vad:
            non_silent_intervals = librosa.effects.split(y, top_db=25)
            if non_silent_intervals.size == 0:
                logging.warning(f"No speech detected in {os.path.basename(file_path)}")
                return np.array([])
            y = np.concatenate([y[start:end] for start, end in non_silent_intervals])

        # 2. MFCC Extraction
        mfccs = librosa.feature.mfcc(y=y, sr=sr, n_mfcc=n_mfcc)

        # 3. Feature Normalization (CMVN)
        mfccs = (mfccs - np.mean(mfccs, axis=1, keepdims=True)) / (np.std(mfccs, axis=1, keepdims=True) + 1e-8)

        # 4. Delta and Delta-Delta Features
        if use_deltas:
            delta1 = librosa.feature.delta(mfccs, order=1)
            delta2 = librosa.feature.delta(mfccs, order=2)
            features = np.vstack([mfccs, delta1, delta2])
        else:
            features = mfccs

        return features.T
    except Exception as e:
        logging.error(f"Error processing {file_path}: {e}")
        return np.array([])

def load_data(data_root: str, max_clips_per_speaker: int = None, **feature_params):
    """Loads data and extracts features for all speakers."""
    data = {}
    speaker_paths = [os.path.join(data_root, spk) for spk in sorted(os.listdir(data_root))]
    speaker_paths = [p for p in speaker_paths if os.path.isdir(p)]

    for spk_path in tqdm(speaker_paths, desc="Loading data"):
        spk = os.path.basename(spk_path)
        files = [f for f in os.listdir(spk_path) if f.lower().endswith('.wav')]
        if max_clips_per_speaker:
            files = files[:max_clips_per_speaker]

        feats_list = [extract_features(os.path.join(spk_path, f), **feature_params) for f in files]
        # Filter out any empty feature arrays from failed extractions
        feats_list = [f for f in feats_list if f.shape[0] > 0]

        if feats_list:
            data[spk] = np.vstack(feats_list)
    return data


# ================================
# 2. Train GMMs
# ================================
def train_gmms(features: dict, n_mix: int) -> dict:
    """Trains a GMM for each speaker."""
    gmms = {}
    for spk, feats in tqdm(features.items(), desc=f"Training GMMs (n_mix={n_mix})"):
        n_frames = feats.shape[0]

        # Adjust mixture count if data is insufficient
        if n_frames < n_mix:
            effective_n_mix = max(1, n_frames)
            logging.warning(f"{spk}: Not enough data ({n_frames} frames). Using {effective_n_mix} components instead of {n_mix}")
        else:
            effective_n_mix = n_mix

        try:
            gmm = GaussianMixture(
                n_components=effective_n_mix,
                covariance_type='diag',
                reg_covar=1e-4,
                max_iter=200,
                n_init=3,
                random_state=0
            )
            gmm.fit(feats)
            gmms[spk] = gmm
        except ValueError as e:
            logging.error(f"Skipping {spk}: GMM fitting failed. {e}")
            continue
    return gmms

# ================================
# 3. Advanced Evaluation
# ================================
def evaluate_eer(gmms: dict, features: dict) -> tuple:
    """
    Calculates the Equal Error Rate (EER) and generates scores for DET curve.
    """
    speakers = list(gmms.keys())
    scores_true, scores_false = [], []

    for spk in tqdm(speakers, desc="Evaluating EER"):
        if spk not in features: continue
        gmm_true = gmms[spk]
        feats_true = features[spk]
        # Score per frame
        scores_true.extend(gmm_true.score_samples(feats_true))

        # Imposter trials
        for other_spk in speakers:
            if other_spk == spk or other_spk not in features: continue
            feats_other = features[other_spk]
            scores_false.extend(gmm_true.score_samples(feats_other))

    if not scores_true or not scores_false:
        logging.error("Evaluation failed: No true or false scores were generated.")
        return 1.0, None, None, None

    y_true = np.array([1] * len(scores_true) + [0] * len(scores_false))
    y_score = np.array(scores_true + scores_false)

    fpr, tpr, thresholds = roc_curve(y_true, y_score)
    fnr = 1 - tpr
    eer_idx = np.nanargmin(np.abs(fnr - fpr))
    eer = (fpr[eer_idx] + fnr[eer_idx]) / 2

    return eer, fpr, fnr, thresholds

def plot_det_curve(eers_data: dict, title: str, filename: str):
    """Plots DET curves for different experiment conditions."""
    plt.figure(figsize=(8, 8))
    for label, (fpr, fnr) in eers_data.items():
        if fpr is not None and fnr is not None:
            plt.plot(fpr * 100, fnr * 100, label=f"{label} (EER: {np.nanmin(np.abs(fnr - fpr))*100:.2f}%)")
    plt.title(title)
    plt.xlabel("False Positive Rate (%)")
    plt.ylabel("False Negative Rate (%)")
    plt.legend()
    plt.grid(True, which='both', linestyle='--')
    plt.savefig(filename)
    plt.close()
    logging.info(f"Saved DET plot to {filename}")

def evaluate_and_plot_confusion_matrix(gmms: dict, features: dict, filename: str):
    """
    Evaluates speaker identification performance and plots a confusion matrix.
    """
    speakers = list(gmms.keys())
    y_true, y_pred = [], []

    for i, true_spk in enumerate(tqdm(speakers, desc="Evaluating Identification")):
        if true_spk not in features: continue
        feats = features[true_spk]

        # To get a single prediction per utterance, we can average frame scores
        # Here we simplify by predicting for the whole concatenated feature block
        log_likelihoods = [gmms[pred_spk].score(feats) for pred_spk in speakers if pred_spk in gmms]

        if not log_likelihoods: continue

        pred_idx = np.argmax(log_likelihoods)

        y_true.append(i)
        y_pred.append(pred_idx)

    if not y_true:
        logging.error("Could not generate predictions for confusion matrix.")
        return

    cm = confusion_matrix(y_true, y_pred, normalize='true')
    disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=speakers)
    fig, ax = plt.subplots(figsize=(10, 10))
    disp.plot(ax=ax, xticks_rotation='vertical', cmap='viridis')
    plt.title("Speaker Identification Confusion Matrix")
    plt.tight_layout()
    plt.savefig(filename)
    plt.close()
    logging.info(f"Saved confusion matrix to {filename}")

# ================================
# 5. Main Experiment Runner
# ================================
def main(args):
    """Main function to run the experiments."""
    logging.info("Starting speaker recognition experiment")

    # Define feature extraction parameters
    feature_params = {'n_mfcc': args.n_mfcc, 'use_deltas': not args.no_deltas, 'use_vad': not args.no_vad}

    # Load full dataset with advanced features
    full_data = load_data(args.data_root, **feature_params)
    if not full_data:
        logging.error("No data loaded. Exiting.")
        return
    logging.info(f"Loaded data for {len(full_data)} speakers. Feature dimension: {next(iter(full_data.values())).shape[1]}")

    # --- Experiment (i): Effect of Unbalanced Dataset ---
    logging.info("\n(i) Testing effect of unbalanced dataset")
    min_len = min(len(v) for v in full_data.values()) if full_data else 0
    balanced_data = {k: v[:min_len] for k, v in full_data.items()}

    gmms_bal = train_gmms(balanced_data, args.n_mix_unbalanced)
    gmms_unb = train_gmms(full_data, args.n_mix_unbalanced)

    eer_bal, fpr_bal, fnr_bal, _ = evaluate_eer(gmms_bal, balanced_data)
    eer_unb, fpr_unb, fnr_unb, _ = evaluate_eer(gmms_unb, full_data)
    logging.info(f"Balanced EER = {eer_bal*100:.2f}% | Unbalanced EER = {eer_unb*100:.2f}%")
    plot_det_curve(
        {"Balanced": (fpr_bal, fnr_bal), "Unbalanced": (fpr_unb, fnr_unb)},
        "DET Curve: Balanced vs Unbalanced Data",
        "det_unbalanced_effect.png"
    )

    # --- Experiment (ii): Effect of Number of Mixture Components ---
    logging.info("\n(ii) Effect of number of mixture components")
    n_components_list = [int(n) for n in args.n_mix_list.split(',')]
    eers_mix = []
    det_data_mix = {}
    for n_mix in n_components_list:
        gmms = train_gmms(full_data, n_mix)
        eer, fpr, fnr, _ = evaluate_eer(gmms, full_data)
        eers_mix.append(eer)
        det_data_mix[f"n_mix={n_mix}"] = (fpr, fnr)
        logging.info(f"n_mix={n_mix}, EER={eer*100:.2f}%")

    plt.figure()
    plt.plot(n_components_list, [e * 100 for e in eers_mix], marker='o')
    plt.title("Effect of Number of Mixture Components on EER")
    plt.xlabel("Number of GMM Components")
    plt.ylabel("Equal Error Rate (EER) [%]")
    plt.grid(True)
    plt.savefig("eer_vs_n_components.png")
    plt.close()
    logging.info("Saved EER vs Components plot to eer_vs_n_components.png")
    plot_det_curve(det_data_mix, "DET Curve: Effect of Mixture Components", "det_mixture_effect.png")

    # --- Experiment (iii): Amount of Data vs Number of Components ---
    logging.info("\n(iii) Amount of data vs number of mixture components")
    clip_counts = [int(c) for c in args.clip_counts.split(',')]
    plt.figure(figsize=(10, 7))
    for clips in clip_counts:
        logging.info(f"Running experiment for {clips} clips per speaker")
        subset_data = load_data(args.data_root, max_clips_per_speaker=clips, **feature_params)
        eers_for_clips = []
        if not subset_data:
            logging.warning(f"No data for {clips} clips, skipping.")
            continue
        for n_mix in n_components_list:
            gmms = train_gmms(subset_data, n_mix)
            eer, _, _, _ = evaluate_eer(gmms, subset_data)
            eers_for_clips.append(eer)
            logging.info(f"clips={clips}, components={n_mix}, EER={eer*100:.2f}%")
        plt.plot(n_components_list, [e * 100 for e in eers_for_clips], marker='o', label=f"{clips} clips/speaker")

    plt.title("EER vs Mixture Components for Different Amounts of Training Data")
    plt.xlabel("Number of GMM Components")
    plt.ylabel("Equal Error Rate (EER) [%]")
    plt.legend()
    plt.grid(True)
    plt.savefig("eer_vs_data_and_components.png")
    plt.close()
    logging.info("Saved EER vs Data/Components plot to eer_vs_data_and_components.png")

    # --- Final Evaluation: Confusion Matrix ---
    logging.info("\n(iv) Generating final confusion matrix")
    # Use the GMMs from the unbalanced experiment as an example
    if gmms_unb and full_data:
        evaluate_and_plot_confusion_matrix(gmms_unb, full_data, "confusion_matrix.png")

    logging.info("\nâœ… Experiment completed successfully!")


if __name__ == "__main__":
    parser = argparse.ArgumentParser(description="Advanced GMM-based Speaker Recognition Experiment")
    parser.add_argument("--data_root", type=str, default="data/data_clips",
                        help="Root directory of the speaker data.")
    parser.add_argument("--n_mfcc", type=int, default=20,
                        help="Number of MFCCs to extract.")
    parser.add_argument("--no_deltas", action='store_true',
                        help="Disable delta and delta-delta features.")
    parser.add_argument("--no_vad", action='store_true',
                        help="Disable Voice Activity Detection.")
    parser.add_argument("--n_mix_unbalanced", type=int, default=8,
                        help="Number of mixture components for the unbalanced data experiment.")
    parser.add_argument("--n_mix_list", type=str, default="2,4,8,16",
                        help="Comma-separated list of mixture components to test.")
    parser.add_argument("--clip_counts", type=str, default="2,4,8,12",
                        help="Comma-separated list of clip counts per speaker to test.")

    # Use parse_known_args() to avoid errors in environments like Jupyter
    args, unknown = parser.parse_known_args()
    main(args)

```

```python
import os
import numpy as np
import librosa
import matplotlib.pyplot as plt
from sklearn.mixture import GaussianMixture
from sklearn.metrics import roc_curve, confusion_matrix, ConfusionMatrixDisplay
from tqdm import tqdm
import argparse
import logging
import warnings

# Suppress warnings from librosa
warnings.filterwarnings('ignore', category=UserWarning, module='librosa')

# ================================
# Setup Logging
# ================================
logging.basicConfig(level=logging.INFO,
                    format='%(asctime)s - %(levelname)s - %(message)s')

# ================================
# 1. Advanced Data Loading & Feature Extraction
# ================================
def extract_features(file_path: str, n_mfcc: int = 20, use_deltas: bool = True, use_vad: bool = True) -> np.ndarray:
    """
    Extracts advanced features from a single audio file.

    Args:
        file_path (str): Path to the audio file.
        n_mfcc (int): Number of MFCCs to compute.
        use_deltas (bool): Whether to append delta and delta-delta features.
        use_vad (bool): Whether to apply Voice Activity Detection.

    Returns:
        np.ndarray: The extracted features, or an empty array if processing fails.
    """
    try:
        y, sr = librosa.load(file_path, sr=None)

        # 1. Voice Activity Detection (VAD)
        if use_vad:
            non_silent_intervals = librosa.effects.split(y, top_db=25)
            if non_silent_intervals.size == 0:
                logging.warning(f"No speech detected in {os.path.basename(file_path)}")
                return np.array([])
            y = np.concatenate([y[start:end] for start, end in non_silent_intervals])

        # 2. MFCC Extraction
        mfccs = librosa.feature.mfcc(y=y, sr=sr, n_mfcc=n_mfcc)

        # 3. Feature Normalization (CMVN)
        mfccs = (mfccs - np.mean(mfccs, axis=1, keepdims=True)) / (np.std(mfccs, axis=1, keepdims=True) + 1e-8)

        # 4. Delta and Delta-Delta Features
        if use_deltas:
            delta1 = librosa.feature.delta(mfccs, order=1)
            delta2 = librosa.feature.delta(mfccs, order=2)
            features = np.vstack([mfccs, delta1, delta2])
        else:
            features = mfccs

        return features.T
    except Exception as e:
        logging.error(f"Error processing {file_path}: {e}")
        return np.array([])

def load_data(data_root: str, max_clips_per_speaker: int = None, verbose: bool = True, **feature_params):
    """Loads data and extracts features for all speakers."""
    data = {}
    speaker_paths = [os.path.join(data_root, spk) for spk in sorted(os.listdir(data_root))]
    speaker_paths = [p for p in speaker_paths if os.path.isdir(p)]

    iterator = tqdm(speaker_paths, desc="Loading data") if verbose else speaker_paths

    for spk_path in iterator:
        spk = os.path.basename(spk_path)
        files = [f for f in os.listdir(spk_path) if f.lower().endswith('.wav')]
        if max_clips_per_speaker:
            files = files[:max_clips_per_speaker]

        feats_list = [extract_features(os.path.join(spk_path, f), **feature_params) for f in files]
        # Filter out any empty feature arrays from failed extractions
        feats_list = [f for f in feats_list if f.shape[0] > 0]

        if feats_list:
            data[spk] = np.vstack(feats_list)
    return data


# ================================
# 2. Train GMMs
# ================================
def train_gmms(features: dict, n_mix: int, verbose: bool = True) -> dict:
    """Trains a GMM for each speaker."""
    gmms = {}

    iterator = tqdm(features.items(), desc=f"Training GMMs (n_mix={n_mix})") if verbose else features.items()

    for spk, feats in iterator:
        n_frames = feats.shape[0]

        # Adjust mixture count if data is insufficient
        if n_frames < n_mix:
            effective_n_mix = max(1, n_frames)
            logging.warning(f"{spk}: Not enough data ({n_frames} frames). Using {effective_n_mix} components instead of {n_mix}")
        else:
            effective_n_mix = n_mix

        try:
            gmm = GaussianMixture(
                n_components=effective_n_mix,
                covariance_type='diag',
                reg_covar=1e-4,
                max_iter=200,
                n_init=3,
                random_state=0
            )
            gmm.fit(feats)
            gmms[spk] = gmm
        except ValueError as e:
            logging.error(f"Skipping {spk}: GMM fitting failed. {e}")
            continue
    return gmms

# ================================
# 3. Advanced Evaluation
# ================================
def evaluate_eer(gmms: dict, features: dict, verbose: bool = True) -> tuple:
    """
    Calculates the Equal Error Rate (EER) and generates scores for DET curve.
    """
    speakers = list(gmms.keys())
    scores_true, scores_false = [], []

    iterator = tqdm(speakers, desc="Evaluating EER") if verbose else speakers

    for spk in iterator:
        if spk not in features or spk not in gmms: continue
        gmm_true = gmms[spk]
        feats_true = features[spk]
        # Score per frame
        scores_true.extend(gmm_true.score_samples(feats_true))

        # Imposter trials
        for other_spk in speakers:
            if other_spk == spk or other_spk not in features: continue
            feats_other = features[other_spk]
            scores_false.extend(gmm_true.score_samples(feats_other))

    if not scores_true or not scores_false:
        logging.error("Evaluation failed: No true or false scores were generated.")
        return 1.0, None, None, None

    y_true = np.array([1] * len(scores_true) + [0] * len(scores_false))
    y_score = np.array(scores_true + scores_false)

    fpr, tpr, thresholds = roc_curve(y_true, y_score)
    fnr = 1 - tpr
    eer_idx = np.nanargmin(np.abs(fnr - fpr))
    eer = (fpr[eer_idx] + fnr[eer_idx]) / 2

    return eer, fpr, fnr, thresholds

def plot_det_curve(eers_data: dict, title: str, filename: str):
    """Plots DET curves for different experiment conditions."""
    plt.figure(figsize=(8, 8))
    for label, (fpr, fnr) in eers_data.items():
        if fpr is not None and fnr is not None:
            plt.plot(fpr * 100, fnr * 100, label=f"{label} (EER: {np.nanmin(np.abs(fnr - fpr))*100:.2f}%)")
    plt.title(title)
    plt.xlabel("False Positive Rate (%)")
    plt.ylabel("False Negative Rate (%)")
    plt.legend()
    plt.grid(True, which='both', linestyle='--')
    logging.info(f"Displaying DET plot: {title}")
    plt.show()

def evaluate_and_plot_confusion_matrix(gmms: dict, features: dict, filename: str, verbose: bool = True):
    """
    Evaluates speaker identification performance and plots a confusion matrix.
    """
    speakers = list(gmms.keys())
    y_true, y_pred = [], []

    iterator = tqdm(speakers, desc="Evaluating Identification") if verbose else speakers

    for i, true_spk in enumerate(iterator):
        if true_spk not in features: continue
        feats = features[true_spk]

        log_likelihoods = [gmms[pred_spk].score(feats) for pred_spk in speakers if pred_spk in gmms]

        if not log_likelihoods: continue

        pred_idx = np.argmax(log_likelihoods)

        y_true.append(i)
        y_pred.append(pred_idx)

    if not y_true:
        logging.error("Could not generate predictions for confusion matrix.")
        return

    cm = confusion_matrix(y_true, y_pred, normalize='true')
    disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=speakers)
    fig, ax = plt.subplots(figsize=(10, 10))
    disp.plot(ax=ax, xticks_rotation='vertical', cmap='viridis')
    plt.title("Speaker Identification Confusion Matrix")
    plt.tight_layout()
    logging.info("Displaying confusion matrix.")
    plt.show()

# ================================
# 5. Main Experiment Runner
# ================================
def main(args):
    """Main function to run the experiments."""
    logging.info("Starting speaker recognition experiment")

    logging.info("="*30)
    logging.info("Experiment Configuration:")
    logging.info(f"  Data Root: {args.data_root}")
    logging.info(f"  MFCCs: {args.n_mfcc}")
    logging.info(f"  Use Deltas: {not args.no_deltas}")
    logging.info(f"  Use VAD: {not args.no_vad}")
    logging.info(f"  Mixture Components (Unbalanced): {args.n_mix_unbalanced}")
    logging.info(f"  Mixture Components List: {args.n_mix_list}")
    logging.info(f"  Clip Counts List: {args.clip_counts}")
    logging.info("="*30)

    # Define feature extraction parameters
    feature_params = {'n_mfcc': args.n_mfcc, 'use_deltas': not args.no_deltas, 'use_vad': not args.no_vad}

    # Load full dataset with advanced features
    full_data = load_data(args.data_root, **feature_params)
    if not full_data:
        logging.error("No data loaded. Exiting.")
        return
    logging.info(f"Loaded data for {len(full_data)} speakers. Feature dimension: {next(iter(full_data.values())).shape[1]}")

    # --- Experiment (i): Effect of Unbalanced Dataset ---
    logging.info("\n(i) Testing effect of unbalanced dataset")
    min_len = min(len(v) for v in full_data.values()) if full_data else 0
    balanced_data = {k: v[:min_len] for k, v in full_data.items()}

    gmms_bal = train_gmms(balanced_data, args.n_mix_unbalanced)
    gmms_unb = train_gmms(full_data, args.n_mix_unbalanced)

    eer_bal, fpr_bal, fnr_bal, _ = evaluate_eer(gmms_bal, balanced_data)
    eer_unb, fpr_unb, fnr_unb, _ = evaluate_eer(gmms_unb, full_data)
    logging.info(f"Balanced EER = {eer_bal*100:.2f}% | Unbalanced EER = {eer_unb*100:.2f}%")
    plot_det_curve(
        {"Balanced": (fpr_bal, fnr_bal), "Unbalanced": (fpr_unb, fnr_unb)},
        "DET Curve: Balanced vs Unbalanced Data",
        "det_unbalanced_effect.png"
    )

    # --- Experiment (ii): Effect of Number of Mixture Components ---
    logging.info("\n(ii) Effect of number of mixture components")
    n_components_list = [int(n) for n in args.n_mix_list.split(',')]
    eers_mix = []
    det_data_mix = {}
    for n_mix in n_components_list:
        gmms = train_gmms(full_data, n_mix)
        eer, fpr, fnr, _ = evaluate_eer(gmms, full_data)
        eers_mix.append(eer)
        det_data_mix[f"n_mix={n_mix}"] = (fpr, fnr)
        logging.info(f"n_mix={n_mix}, EER={eer*100:.2f}%")

    plt.figure()
    plt.plot(n_components_list, [e * 100 for e in eers_mix], marker='o')
    plt.title("Effect of Number of Mixture Components on EER")
    plt.xlabel("Number of GMM Components")
    plt.ylabel("Equal Error Rate (EER) [%]")
    plt.grid(True)
    logging.info("Displaying EER vs Components plot.")
    plt.show()
    plot_det_curve(det_data_mix, "DET Curve: Effect of Mixture Components", "det_mixture_effect.png")

    # --- Experiment (iii): Amount of Data vs Number of Components ---
    logging.info("\n(iii) Amount of data vs number of mixture components")
    clip_counts = [int(c) for c in args.clip_counts.split(',')]
    plt.figure(figsize=(10, 7))
    for clips in clip_counts:
        logging.info(f"Running experiment for {clips} clips per speaker")
        # Disable verbose progress bars for this repetitive section
        subset_data = load_data(args.data_root, max_clips_per_speaker=clips, verbose=False, **feature_params)
        eers_for_clips = []
        if not subset_data:
            logging.warning(f"No data for {clips} clips, skipping.")
            continue
        for n_mix in n_components_list:
            gmms = train_gmms(subset_data, n_mix, verbose=False)
            eer, _, _, _ = evaluate_eer(gmms, subset_data, verbose=False)
            eers_for_clips.append(eer)
            logging.info(f"clips={clips}, components={n_mix}, EER={eer*100:.2f}%")
        plt.plot(n_components_list, [e * 100 for e in eers_for_clips], marker='o', label=f"{clips} clips/speaker")

    plt.title("EER vs Mixture Components for Different Amounts of Training Data")
    plt.xlabel("Number of GMM Components")
    plt.ylabel("Equal Error Rate (EER) [%]")
    plt.legend()
    plt.grid(True)
    logging.info("Displaying EER vs Data/Components plot.")
    plt.show()

    # --- Final Evaluation: Confusion Matrix ---
    logging.info("\n(iv) Generating final confusion matrix")
    # Use the GMMs from the unbalanced experiment as an example
    if gmms_unb and full_data:
        evaluate_and_plot_confusion_matrix(gmms_unb, full_data, "confusion_matrix.png")

    logging.info("\nâœ… Experiment completed successfully!")


if __name__ == "__main__":
    parser = argparse.ArgumentParser(description="Advanced GMM-based Speaker Recognition Experiment")
    parser.add_argument("--data_root", type=str, default="data/data_clips",
                        help="Root directory of the speaker data.")
    parser.add_argument("--n_mfcc", type=int, default=20,
                        help="Number of MFCCs to extract.")
    parser.add_argument("--no_deltas", action='store_true',
                        help="Disable delta and delta-delta features.")
    parser.add_argument("--no_vad", action='store_true',
                        help="Disable Voice Activity Detection.")
    parser.add_argument("--n_mix_unbalanced", type=int, default=8,
                        help="Number of mixture components for the unbalanced data experiment.")
    parser.add_argument("--n_mix_list", type=str, default="2,4,8,16",
                        help="Comma-separated list of mixture components to test.")
    parser.add_argument("--clip_counts", type=str, default="2,4,8,12",
                        help="Comma-separated list of clip counts per speaker to test.")

    # Use parse_known_args() to avoid errors in environments like Jupyter
    args, unknown = parser.parse_known_args()
    main(args)
```

```python
import os
import numpy as np
import librosa
import matplotlib.pyplot as plt
from sklearn.mixture import GaussianMixture
from sklearn.metrics import roc_curve, confusion_matrix, ConfusionMatrixDisplay
from tqdm import tqdm
import argparse
import logging
import warnings

# Suppress warnings from librosa
warnings.filterwarnings('ignore', category=UserWarning, module='librosa')

# ================================
# Setup Logging
# ================================
logging.basicConfig(level=logging.INFO,
                    format='%(asctime)s - %(levelname)s - %(message)s')

# ================================
# 1. Advanced Data Loading & Feature Extraction
# ================================
def extract_features(file_path: str, n_mfcc: int = 20, use_deltas: bool = True, use_vad: bool = True) -> np.ndarray:
    """
    Extracts advanced features from a single audio file.

    Args:
        file_path (str): Path to the audio file.
        n_mfcc (int): Number of MFCCs to compute.
        use_deltas (bool): Whether to append delta and delta-delta features.
        use_vad (bool): Whether to apply Voice Activity Detection.

    Returns:
        np.ndarray: The extracted features, or an empty array if processing fails.
    """
    try:
        y, sr = librosa.load(file_path, sr=None)

        # 1. Voice Activity Detection (VAD)
        if use_vad:
            non_silent_intervals = librosa.effects.split(y, top_db=25)
            if non_silent_intervals.size == 0:
                logging.warning(f"No speech detected in {os.path.basename(file_path)}")
                return np.array([])
            y = np.concatenate([y[start:end] for start, end in non_silent_intervals])

        # 2. MFCC Extraction
        mfccs = librosa.feature.mfcc(y=y, sr=sr, n_mfcc=n_mfcc)

        # 3. Feature Normalization (CMVN)
        mfccs = (mfccs - np.mean(mfccs, axis=1, keepdims=True)) / (np.std(mfccs, axis=1, keepdims=True) + 1e-8)

        # 4. Delta and Delta-Delta Features
        if use_deltas:
            # The default width for librosa.feature.delta is 9 frames.
            if mfccs.shape[1] < 9:
                logging.warning(f"File {os.path.basename(file_path)} is too short ({mfccs.shape[1]} frames) to compute delta features. Skipping file.")
                return np.array([])

            delta1 = librosa.feature.delta(mfccs, order=1)
            delta2 = librosa.feature.delta(mfccs, order=2)
            features = np.vstack([mfccs, delta1, delta2])
        else:
            features = mfccs

        return features.T
    except Exception as e:
        logging.error(f"Error processing {file_path}: {e}")
        return np.array([])

def load_data(data_root: str, max_clips_per_speaker: int = None, verbose: bool = True, **feature_params):
    """Loads data and extracts features for all speakers."""
    data = {}
    speaker_paths = [os.path.join(data_root, spk) for spk in sorted(os.listdir(data_root))]
    speaker_paths = [p for p in speaker_paths if os.path.isdir(p)]

    iterator = tqdm(speaker_paths, desc="Loading data") if verbose else speaker_paths

    for spk_path in iterator:
        spk = os.path.basename(spk_path)
        files = [f for f in os.listdir(spk_path) if f.lower().endswith('.wav')]
        if max_clips_per_speaker:
            files = files[:max_clips_per_speaker]

        feats_list = [extract_features(os.path.join(spk_path, f), **feature_params) for f in files]
        # Filter out any empty feature arrays from failed extractions
        feats_list = [f for f in feats_list if f.shape[0] > 0]

        if feats_list:
            data[spk] = np.vstack(feats_list)
    return data


# ================================
# 2. Train GMMs
# ================================
def train_gmms(features: dict, n_mix: int, verbose: bool = True) -> dict:
    """Trains a GMM for each speaker."""
    gmms = {}

    iterator = tqdm(features.items(), desc=f"Training GMMs (n_mix={n_mix})") if verbose else features.items()

    for spk, feats in iterator:
        n_frames = feats.shape[0]

        # Adjust mixture count if data is insufficient
        if n_frames < n_mix:
            effective_n_mix = max(1, n_frames)
            logging.warning(f"{spk}: Not enough data ({n_frames} frames). Using {effective_n_mix} components instead of {n_mix}")
        else:
            effective_n_mix = n_mix

        try:
            gmm = GaussianMixture(
                n_components=effective_n_mix,
                covariance_type='diag',
                reg_covar=1e-4,
                max_iter=200,
                n_init=3,
                random_state=0
            )
            gmm.fit(feats)
            gmms[spk] = gmm
        except ValueError as e:
            logging.error(f"Skipping {spk}: GMM fitting failed. {e}")
            continue
    return gmms

# ================================
# 3. Advanced Evaluation
# ================================
def evaluate_eer(gmms: dict, features: dict, verbose: bool = True) -> tuple:
    """
    Calculates the Equal Error Rate (EER) and generates scores for DET curve.
    """
    speakers = list(gmms.keys())
    scores_true, scores_false = [], []

    iterator = tqdm(speakers, desc="Evaluating EER") if verbose else speakers

    for spk in iterator:
        if spk not in features or spk not in gmms: continue
        gmm_true = gmms[spk]
        feats_true = features[spk]
        # Score per frame
        scores_true.extend(gmm_true.score_samples(feats_true))

        # Imposter trials
        for other_spk in speakers:
            if other_spk == spk or other_spk not in features: continue
            feats_other = features[other_spk]
            scores_false.extend(gmm_true.score_samples(feats_other))

    if not scores_true or not scores_false:
        logging.error("Evaluation failed: No true or false scores were generated.")
        return 1.0, None, None, None

    y_true = np.array([1] * len(scores_true) + [0] * len(scores_false))
    y_score = np.array(scores_true + scores_false)

    fpr, tpr, thresholds = roc_curve(y_true, y_score)
    fnr = 1 - tpr
    eer_idx = np.nanargmin(np.abs(fnr - fpr))
    eer = (fpr[eer_idx] + fnr[eer_idx]) / 2

    return eer, fpr, fnr, thresholds

def plot_det_curve(eers_data: dict, title: str, filename: str):
    """Plots DET curves for different experiment conditions."""
    plt.figure(figsize=(8, 8))
    for label, (fpr, fnr) in eers_data.items():
        if fpr is not None and fnr is not None:
            plt.plot(fpr * 100, fnr * 100, label=f"{label} (EER: {np.nanmin(np.abs(fnr - fpr))*100:.2f}%)")
    plt.title(title)
    plt.xlabel("False Positive Rate (%)")
    plt.ylabel("False Negative Rate (%)")
    plt.legend()
    plt.grid(True, which='both', linestyle='--')
    logging.info(f"Displaying DET plot: {title}")
    plt.show()

def evaluate_and_plot_confusion_matrix(gmms: dict, features: dict, filename: str, verbose: bool = True):
    """
    Evaluates speaker identification performance and plots a confusion matrix.
    """
    speakers = list(gmms.keys())
    y_true, y_pred = [], []

    iterator = tqdm(speakers, desc="Evaluating Identification") if verbose else speakers

    for i, true_spk in enumerate(iterator):
        if true_spk not in features: continue
        feats = features[true_spk]

        log_likelihoods = [gmms[pred_spk].score(feats) for pred_spk in speakers if pred_spk in gmms]

        if not log_likelihoods: continue

        pred_idx = np.argmax(log_likelihoods)

        y_true.append(i)
        y_pred.append(pred_idx)

    if not y_true:
        logging.error("Could not generate predictions for confusion matrix.")
        return

    cm = confusion_matrix(y_true, y_pred, normalize='true')
    disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=speakers)
    fig, ax = plt.subplots(figsize=(10, 10))
    disp.plot(ax=ax, xticks_rotation='vertical', cmap='viridis')
    plt.title("Speaker Identification Confusion Matrix")
    plt.tight_layout()
    logging.info("Displaying confusion matrix.")
    plt.show()

# ================================
# 5. Main Experiment Runner
# ================================
def main(args):
    """Main function to run the experiments."""
    logging.info("Starting speaker recognition experiment")

    logging.info("="*30)
    logging.info("Experiment Configuration:")
    logging.info(f"  Data Root: {args.data_root}")
    logging.info(f"  MFCCs: {args.n_mfcc}")
    logging.info(f"  Use Deltas: {not args.no_deltas}")
    logging.info(f"  Use VAD: {not args.no_vad}")
    logging.info(f"  Mixture Components (Unbalanced): {args.n_mix_unbalanced}")
    logging.info(f"  Mixture Components List: {args.n_mix_list}")
    logging.info(f"  Clip Counts List: {args.clip_counts}")
    logging.info("="*30)

    # Define feature extraction parameters
    feature_params = {'n_mfcc': args.n_mfcc, 'use_deltas': not args.no_deltas, 'use_vad': not args.no_vad}

    # Load full dataset with advanced features
    full_data = load_data(args.data_root, **feature_params)
    if not full_data:
        logging.error("No data loaded. Exiting.")
        return
    logging.info(f"Loaded data for {len(full_data)} speakers. Feature dimension: {next(iter(full_data.values())).shape[1]}")

    # --- Experiment (i): Effect of Unbalanced Dataset ---
    logging.info("\n(i) Testing effect of unbalanced dataset")
    min_len = min(len(v) for v in full_data.values()) if full_data else 0
    balanced_data = {k: v[:min_len] for k, v in full_data.items()}

    gmms_bal = train_gmms(balanced_data, args.n_mix_unbalanced)
    gmms_unb = train_gmms(full_data, args.n_mix_unbalanced)

    eer_bal, fpr_bal, fnr_bal, _ = evaluate_eer(gmms_bal, balanced_data)
    eer_unb, fpr_unb, fnr_unb, _ = evaluate_eer(gmms_unb, full_data)
    logging.info(f"Balanced EER = {eer_bal*100:.2f}% | Unbalanced EER = {eer_unb*100:.2f}%")
    plot_det_curve(
        {"Balanced": (fpr_bal, fnr_bal), "Unbalanced": (fpr_unb, fnr_unb)},
        "DET Curve: Balanced vs Unbalanced Data",
        "det_unbalanced_effect.png"
    )

    # --- Experiment (ii): Effect of Number of Mixture Components ---
    logging.info("\n(ii) Effect of number of mixture components")
    n_components_list = [int(n) for n in args.n_mix_list.split(',')]
    eers_mix = []
    det_data_mix = {}
    for n_mix in n_components_list:
        gmms = train_gmms(full_data, n_mix)
        eer, fpr, fnr, _ = evaluate_eer(gmms, full_data)
        eers_mix.append(eer)
        det_data_mix[f"n_mix={n_mix}"] = (fpr, fnr)
        logging.info(f"n_mix={n_mix}, EER={eer*100:.2f}%")

    plt.figure()
    plt.plot(n_components_list, [e * 100 for e in eers_mix], marker='o')
    plt.title("Effect of Number of Mixture Components on EER")
    plt.xlabel("Number of GMM Components")
    plt.ylabel("Equal Error Rate (EER) [%]")
    plt.grid(True)
    logging.info("Displaying EER vs Components plot.")
    plt.show()
    plot_det_curve(det_data_mix, "DET Curve: Effect of Mixture Components", "det_mixture_effect.png")

    # --- Experiment (iii): Amount of Data vs Number of Components ---
    logging.info("\n(iii) Amount of data vs number of mixture components")
    clip_counts = [int(c) for c in args.clip_counts.split(',')]
    plt.figure(figsize=(10, 7))
    for clips in clip_counts:
        logging.info(f"Running experiment for {clips} clips per speaker")
        # Disable verbose progress bars for this repetitive section
        subset_data = load_data(args.data_root, max_clips_per_speaker=clips, verbose=False, **feature_params)
        eers_for_clips = []
        if not subset_data:
            logging.warning(f"No data for {clips} clips, skipping.")
            continue
        for n_mix in n_components_list:
            gmms = train_gmms(subset_data, n_mix, verbose=False)
            eer, _, _, _ = evaluate_eer(gmms, subset_data, verbose=False)
            eers_for_clips.append(eer)
            logging.info(f"clips={clips}, components={n_mix}, EER={eer*100:.2f}%")
        plt.plot(n_components_list, [e * 100 for e in eers_for_clips], marker='o', label=f"{clips} clips/speaker")

    plt.title("EER vs Mixture Components for Different Amounts of Training Data")
    plt.xlabel("Number of GMM Components")
    plt.ylabel("Equal Error Rate (EER) [%]")
    plt.legend()
    plt.grid(True)
    logging.info("Displaying EER vs Data/Components plot.")
    plt.show()

    # --- Final Evaluation: Confusion Matrix ---
    logging.info("\n(iv) Generating final confusion matrix")
    # Use the GMMs from the unbalanced experiment as an example
    if gmms_unb and full_data:
        evaluate_and_plot_confusion_matrix(gmms_unb, full_data, "confusion_matrix.png")

    logging.info("\nâœ… Experiment completed successfully!")


if __name__ == "__main__":
    parser = argparse.ArgumentParser(description="Advanced GMM-based Speaker Recognition Experiment")
    parser.add_argument("--data_root", type=str, default="data/data_clips",
                        help="Root directory of the speaker data.")
    parser.add_argument("--n_mfcc", type=int, default=20,
                        help="Number of MFCCs to extract.")
    parser.add_argument("--no_deltas", action='store_true',
                        help="Disable delta and delta-delta features.")
    parser.add_argument("--no_vad", action='store_true',
                        help="Disable Voice Activity Detection.")
    parser.add_argument("--n_mix_unbalanced", type=int, default=8,
                        help="Number of mixture components for the unbalanced data experiment.")
    parser.add_argument("--n_mix_list", type=str, default="2,4,8,16",
                        help="Comma-separated list of mixture components to test.")
    parser.add_argument("--clip_counts", type=str, default="2,4,8,12",
                        help="Comma-separated list of clip counts per speaker to test.")

    # Use parse_known_args() to avoid errors in environments like Jupyter
    args, unknown = parser.parse_known_args()
    main(args)

```

```python
import os
import numpy as np
import librosa
import matplotlib.pyplot as plt
from sklearn.mixture import GaussianMixture
from sklearn.metrics import roc_curve, confusion_matrix, ConfusionMatrixDisplay
from sklearn.model_selection import train_test_split
from tqdm import tqdm
import argparse
import logging
import warnings
import random

# Suppress warnings from librosa
warnings.filterwarnings('ignore', category=UserWarning, module='librosa')

# ================================
# Setup Logging
# ================================
logging.basicConfig(level=logging.INFO,
                    format='%(asctime)s - %(levelname)s - %(message)s')

# ================================
# 1. Advanced Data Loading & Feature Extraction with Train/Test Split
# ================================
def extract_features(file_path: str, n_mfcc: int = 20, use_deltas: bool = True, use_vad: bool = True) -> np.ndarray:
    """
    Extracts advanced features from a single audio file.
    (Function body remains the same as before)
    """
    try:
        y, sr = librosa.load(file_path, sr=None)

        # 1. Voice Activity Detection (VAD)
        if use_vad:
            non_silent_intervals = librosa.effects.split(y, top_db=25)
            if non_silent_intervals.size == 0:
                logging.warning(f"No speech detected in {os.path.basename(file_path)}")
                return np.array([])
            y = np.concatenate([y[start:end] for start, end in non_silent_intervals])

        if len(y) == 0: return np.array([])

        # 2. MFCC Extraction
        mfccs = librosa.feature.mfcc(y=y, sr=sr, n_mfcc=n_mfcc)

        # 3. Feature Normalization (CMVN)
        mfccs = (mfccs - np.mean(mfccs, axis=1, keepdims=True)) / (np.std(mfccs, axis=1, keepdims=True) + 1e-8)

        # 4. Delta and Delta-Delta Features
        if use_deltas:
            if mfccs.shape[1] < 9: # Default width for delta calculation
                logging.warning(f"File {os.path.basename(file_path)} is too short ({mfccs.shape[1]} frames) for deltas. Skipping.")
                return np.array([])
            delta1 = librosa.feature.delta(mfccs, order=1)
            delta2 = librosa.feature.delta(mfccs, order=2)
            features = np.vstack([mfccs, delta1, delta2])
        else:
            features = mfccs

        return features.T
    except Exception as e:
        logging.error(f"Error processing {file_path}: {e}")
        return np.array([])

def load_data_and_split(data_root: str, test_size: float = 0.2, verbose: bool = True, **feature_params):
    """
    Loads data, splits it into training and testing sets, and extracts features.
    """
    train_data, test_data = {}, {}
    speaker_paths = [os.path.join(data_root, spk) for spk in sorted(os.listdir(data_root))]
    speaker_paths = [p for p in speaker_paths if os.path.isdir(p)]

    iterator = tqdm(speaker_paths, desc="Loading data and splitting") if verbose else speaker_paths

    for spk_path in iterator:
        spk = os.path.basename(spk_path)
        files = [os.path.join(spk_path, f) for f in os.listdir(spk_path) if f.lower().endswith('.wav')]

        if len(files) < 2:
            logging.warning(f"Speaker {spk} has fewer than 2 clips, cannot split. Skipping speaker.")
            continue

        train_files, test_files = train_test_split(files, test_size=test_size, random_state=42)

        # Process training files
        train_feats_list = [extract_features(f, **feature_params) for f in train_files]
        train_feats_list = [f for f in train_feats_list if f.shape[0] > 0]
        if train_feats_list:
            train_data[spk] = np.vstack(train_feats_list)

        # Process testing files
        test_feats_list = [extract_features(f, **feature_params) for f in test_files]
        test_feats_list = [f for f in test_feats_list if f.shape[0] > 0]
        if test_feats_list:
            test_data[spk] = np.vstack(test_feats_list)

    return train_data, test_data


# ================================
# 2. Train GMMs
# ================================
def train_gmms(features: dict, n_mix: int, verbose: bool = True) -> dict:
    """
    Trains a GMM for each speaker.
    (Function body remains the same as before)
    """
    gmms = {}
    iterator = tqdm(features.items(), desc=f"Training GMMs (n_mix={n_mix})") if verbose else features.items()
    for spk, feats in iterator:
        n_frames = feats.shape[0]
        effective_n_mix = n_mix
        if n_frames < n_mix:
            effective_n_mix = max(1, n_frames // 10) # Heuristic
            logging.warning(f"{spk}: Not enough data. Using {effective_n_mix} components.")
        try:
            gmm = GaussianMixture(n_components=effective_n_mix, covariance_type='diag', reg_covar=1e-4, max_iter=200, n_init=3, random_state=0)
            gmm.fit(feats)
            gmms[spk] = gmm
        except ValueError as e:
            logging.error(f"Skipping {spk}: GMM fitting failed. {e}")
            continue
    return gmms

# ================================
# 3. Advanced Evaluation (on Test Set)
# ================================
def evaluate_eer(gmms: dict, features_test: dict, verbose: bool = True) -> tuple:
    """
    Calculates EER using trained GMMs on a separate test set.
    """
    speakers = list(gmms.keys())
    scores_true, scores_false = [], []

    iterator = tqdm(speakers, desc="Evaluating EER on test set") if verbose else speakers

    for spk in iterator:
        if spk not in features_test or spk not in gmms: continue
        gmm_true = gmms[spk]
        feats_true = features_test[spk]
        scores_true.extend(gmm_true.score_samples(feats_true))

        # Imposter trials
        for other_spk in speakers:
            if other_spk == spk or other_spk not in features_test: continue
            feats_other = features_test[other_spk]
            scores_false.extend(gmm_true.score_samples(feats_other))

    if not scores_true or not scores_false:
        logging.error("Evaluation failed: No scores generated.")
        return 1.0, None, None, None

    y_true = np.array([1] * len(scores_true) + [0] * len(scores_false))
    y_score = np.array(scores_true + scores_false)

    fpr, tpr, thresholds = roc_curve(y_true, y_score)
    fnr = 1 - tpr
    eer_idx = np.nanargmin(np.abs(fnr - fpr))
    eer = (fpr[eer_idx] + fnr[eer_idx]) / 2

    return eer, fpr, fnr, thresholds

# Plotting functions (plot_det_curve, etc.) remain largely the same
def plot_det_curve(eers_data: dict, title: str):
    plt.figure(figsize=(8, 8))
    for label, (eer, fpr, fnr) in eers_data.items():
        if fpr is not None and fnr is not None:
            plt.plot(fpr * 100, fnr * 100, label=f"{label} (EER: {eer*100:.2f}%)")
    plt.title(title)
    plt.xlabel("False Positive Rate (%)")
    plt.ylabel("False Negative Rate (%)")
    plt.legend()
    plt.grid(True, which='both', linestyle='--')
    logging.info(f"Displaying DET plot: {title}")
    plt.show()

def evaluate_and_plot_confusion_matrix(gmms: dict, features_test: dict, verbose: bool = True):
    speakers = list(gmms.keys())
    y_true_labels, y_pred_labels = [], []

    iterator = tqdm(speakers, desc="Evaluating Identification on test set") if verbose else speakers
    for true_spk in iterator:
        if true_spk not in features_test: continue
        feats = features_test[true_spk]
        log_likelihoods = {pred_spk: gmms[pred_spk].score(feats) for pred_spk in speakers if pred_spk in gmms}
        if not log_likelihoods: continue
        pred_spk = max(log_likelihoods, key=log_likelihoods.get)
        y_true_labels.append(true_spk)
        y_pred_labels.append(pred_spk)

    if not y_true_labels:
        logging.error("Could not generate predictions for confusion matrix.")
        return

    cm = confusion_matrix(y_true_labels, y_pred_labels, labels=speakers, normalize='true')
    disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=speakers)
    fig, ax = plt.subplots(figsize=(10, 10))
    disp.plot(ax=ax, xticks_rotation='vertical', cmap='viridis')
    plt.title("Speaker Identification Confusion Matrix (on Test Set)")
    plt.tight_layout()
    logging.info("Displaying confusion matrix.")
    plt.show()

# ================================
# 5. Main Experiment Runner
# ================================
def main(args):
    """Main function to run the experiments."""
    logging.info("="*30)
    logging.info("Experiment Configuration:")
    logging.info(f"  Data Root: {args.data_root}")
    logging.info(f"  Test Size: {args.test_size}")
    # ... print other args ...
    logging.info("="*30)

    feature_params = {'n_mfcc': args.n_mfcc, 'use_deltas': not args.no_deltas, 'use_vad': not args.no_vad}

    # --- Load data with 80/20 split ---
    train_data, test_data = load_data_and_split(args.data_root, test_size=args.test_size, **feature_params)
    if not train_data or not test_data:
        logging.error("Data loading failed or resulted in empty train/test sets. Exiting.")
        return
    logging.info(f"Data split complete. Training on {len(train_data)} speakers, testing on {len(test_data)} speakers.")

    # --- NEW Experiment (i): Conclusive Effect of Unbalanced Dataset ---
    logging.info("\n(i) Testing conclusive effect of unbalanced dataset")
    imbalance_levels = [0.0, 0.25, 0.50, 0.75] # % of data to remove
    unbalanced_results = {}

    speakers_to_reduce = random.sample(list(train_data.keys()), k=len(train_data) // 2)
    logging.info(f"Will reduce data for speakers: {speakers_to_reduce}")

    for level in imbalance_levels:
        logging.info(f"Creating training set with {level*100}% data reduction.")
        unbalanced_train_data = {}
        for spk, feats in train_data.items():
            if spk in speakers_to_reduce:
                reduce_to_frames = int(feats.shape[0] * (1 - level))
                unbalanced_train_data[spk] = feats[:reduce_to_frames, :]
            else:
                unbalanced_train_data[spk] = feats

        gmms = train_gmms(unbalanced_train_data, args.n_mix_unbalanced, verbose=False)
        eer, fpr, fnr, _ = evaluate_eer(gmms, test_data, verbose=False)
        unbalanced_results[f"{int(level*100)}% reduction"] = (eer, fpr, fnr)
        logging.info(f"Imbalance Level {level*100}% -> EER = {eer*100:.2f}%")

    # Plot the results of the imbalance experiment
    plt.figure(figsize=(10, 7))
    eers = [res[0] * 100 for res in unbalanced_results.values()]
    plt.plot(imbalance_levels, eers, marker='o')
    plt.title("Effect of Data Imbalance on EER")
    plt.xlabel("Percentage of Data Removed for Half of Speakers")
    plt.ylabel("Equal Error Rate (EER) [%]")
    plt.xticks([l for l in imbalance_levels], [f'{int(l*100)}%' for l in imbalance_levels])
    plt.grid(True)
    plt.show()
    plot_det_curve(unbalanced_results, "DET Curve: Effect of Data Imbalance")


    # --- Experiment (ii): Effect of Number of Mixture Components (using proper split) ---
    logging.info("\n(ii) Effect of number of mixture components")
    n_components_list = [int(n) for n in args.n_mix_list.split(',')]
    eers_mix = []
    for n_mix in n_components_list:
        gmms = train_gmms(train_data, n_mix) # Train on training data
        eer, _, _, _ = evaluate_eer(gmms, test_data) # Test on test data
        eers_mix.append(eer)
        logging.info(f"n_mix={n_mix}, EER={eer*100:.2f}%")
    # ... (plotting code remains the same) ...
    plt.figure()
    plt.plot(n_components_list, [e*100 for e in eers_mix], marker='o')
    plt.title("Effect of Mixture Components (Evaluated on Test Set)")
    plt.xlabel("Number of GMM Components")
    plt.ylabel("EER (%)")
    plt.grid(True)
    plt.show()


    # --- Experiment (iii) is implicitly covered by the new unbalancing experiment.
    # The original was "Amount of Data vs. Number of Components", which is
    # fundamentally what the new imbalance experiment explores in a more
    # controlled way. We can remove it to avoid redundancy.

    # --- Final Evaluation: Confusion Matrix (using model from balanced training) ---
    logging.info("\n(iv) Generating final confusion matrix on test set")
    # We train one final model on the full (balanced) training data for this
    gmms_final = train_gmms(train_data, args.n_mix_unbalanced)
    if gmms_final:
        evaluate_and_plot_confusion_matrix(gmms_final, test_data)

    logging.info("\nâœ… Experiment completed successfully!")


if __name__ == "__main__":
    parser = argparse.ArgumentParser(description="Advanced GMM-based Speaker Recognition Experiment")
    # Add test_size argument
    parser.add_argument("--test_size", type=float, default=0.2, help="Proportion of the dataset to use for testing.")
    # Other arguments remain the same
    parser.add_argument("--data_root", type=str, default="data/data_clips", help="Root directory of the speaker data.")
    parser.add_argument("--n_mfcc", type=int, default=20, help="Number of MFCCs to extract.")
    parser.add_argument("--no_deltas", action='store_true', help="Disable delta and delta-delta features.")
    parser.add_argument("--no_vad", action='store_true', help="Disable Voice Activity Detection.")
    parser.add_argument("--n_mix_unbalanced", type=int, default=8, help="Number of mixture components for the unbalanced data experiment.")
    parser.add_argument("--n_mix_list", type=str, default="2,4,8,16", help="Comma-separated list of mixture components to test.")

    args, unknown = parser.parse_known_args()
    main(args)

```

```python
import os
import numpy as np
import librosa
import matplotlib.pyplot as plt
from sklearn.mixture import GaussianMixture
from sklearn.metrics import roc_curve, confusion_matrix, ConfusionMatrixDisplay
from sklearn.model_selection import train_test_split
from tqdm import tqdm
import argparse
import logging
import warnings
import random

# Suppress warnings from librosa
warnings.filterwarnings('ignore', category=UserWarning, module='librosa')

# ================================
# Setup Logging
# ================================
logging.basicConfig(level=logging.INFO,
                    format='%(asctime)s - %(levelname)s - %(message)s')

# ================================
# 1. Advanced Data Loading & Feature Extraction with Train/Test Split
# ================================
def extract_features(file_path: str, n_mfcc: int = 20, use_deltas: bool = True, use_vad: bool = True) -> np.ndarray:
    """
    Extracts advanced features from a single audio file.
    """
    try:
        y, sr = librosa.load(file_path, sr=None)
        if use_vad:
            non_silent_intervals = librosa.effects.split(y, top_db=25)
            if non_silent_intervals.size == 0:
                logging.warning(f"No speech detected in {os.path.basename(file_path)}")
                return np.array([])
            y = np.concatenate([y[start:end] for start, end in non_silent_intervals])
        if len(y) == 0: return np.array([])
        mfccs = librosa.feature.mfcc(y=y, sr=sr, n_mfcc=n_mfcc)
        mfccs = (mfccs - np.mean(mfccs, axis=1, keepdims=True)) / (np.std(mfccs, axis=1, keepdims=True) + 1e-8)
        if use_deltas:
            if mfccs.shape[1] < 9:
                logging.warning(f"File {os.path.basename(file_path)} is too short for deltas. Skipping.")
                return np.array([])
            delta1 = librosa.feature.delta(mfccs, order=1)
            delta2 = librosa.feature.delta(mfccs, order=2)
            features = np.vstack([mfccs, delta1, delta2])
        else:
            features = mfccs
        return features.T
    except Exception as e:
        logging.error(f"Error processing {file_path}: {e}")
        return np.array([])

def load_data_and_split(data_root: str, test_size: float = 0.2, verbose: bool = True, **feature_params):
    """
    Loads data, splits files into training/testing, and extracts features.
    Crucially, it keeps test clips as separate items in a list.
    """
    train_data, test_data = {}, {}
    speaker_paths = [os.path.join(data_root, spk) for spk in sorted(os.listdir(data_root))]
    speaker_paths = [p for p in speaker_paths if os.path.isdir(p)]
    iterator = tqdm(speaker_paths, desc="Loading data and splitting") if verbose else speaker_paths

    for spk_path in iterator:
        spk = os.path.basename(spk_path)
        files = [os.path.join(spk_path, f) for f in os.listdir(spk_path) if f.lower().endswith('.wav')]
        if len(files) < 2:
            logging.warning(f"Speaker {spk} has < 2 clips, cannot split. Skipping.")
            continue
        train_files, test_files = train_test_split(files, test_size=test_size, random_state=42)

        # Training data is concatenated into one big array per speaker
        train_feats_list = [extract_features(f, **feature_params) for f in train_files]
        train_feats_list = [f for f in train_feats_list if f.shape[0] > 0]
        if train_feats_list:
            train_data[spk] = np.vstack(train_feats_list)

        # Test data is kept as a list of arrays (one per clip)
        test_feats_list = [extract_features(f, **feature_params) for f in test_files]
        test_feats_list = [f for f in test_feats_list if f.shape[0] > 0]
        if test_feats_list:
            test_data[spk] = test_feats_list

    return train_data, test_data


# ================================
# 2. Train GMMs
# ================================
def train_gmms(features: dict, n_mix: int, verbose: bool = True) -> dict:
    """
    Trains a GMM for each speaker.
    (No changes needed here)
    """
    gmms = {}
    iterator = tqdm(features.items(), desc=f"Training GMMs (n_mix={n_mix})") if verbose else features.items()
    for spk, feats in iterator:
        n_frames = feats.shape[0]
        effective_n_mix = n_mix
        if n_frames < n_mix * 10: # Heuristic: at least 10 frames per component
            effective_n_mix = max(1, n_frames // 10)
            logging.warning(f"{spk}: Not enough data for {n_mix} components. Using {effective_n_mix}.")
        try:
            gmm = GaussianMixture(n_components=effective_n_mix, covariance_type='diag', reg_covar=1e-4, max_iter=200, n_init=3, random_state=0)
            gmm.fit(feats)
            gmms[spk] = gmm
        except ValueError as e:
            logging.error(f"Skipping {spk}: GMM fitting failed. {e}")
            continue
    return gmms

# ================================
# 3. Advanced Evaluation (on Test Set) - REWRITTEN FOR PER-CLIP EVALUATION
# ================================
def evaluate_eer_per_clip(gmms: dict, features_test: dict, verbose: bool = True) -> tuple:
    """
    Calculates EER on a per-clip basis. `features_test` is a dict where each
    value is a LIST of feature arrays (one per clip).
    """
    speakers = list(gmms.keys())
    scores_true, scores_false = [], []

    iterator = tqdm(speakers, desc="Evaluating EER on test clips") if verbose else speakers

    for true_spk in iterator:
        if true_spk not in features_test or true_spk not in gmms: continue

        # True trials: score clips from the correct speaker
        for clip_feats in features_test[true_spk]:
            scores_true.append(gmms[true_spk].score(clip_feats))

            # Imposter trials: score this clip against all other speakers
            for imposter_spk in speakers:
                if imposter_spk != true_spk and imposter_spk in gmms:
                    scores_false.append(gmms[imposter_spk].score(clip_feats))

    if not scores_true or not scores_false:
        logging.error("Evaluation failed: No scores generated.")
        return 1.0, None, None, None

    y_true = np.array([1] * len(scores_true) + [0] * len(scores_false))
    y_score = np.array(scores_true + scores_false)
    fpr, tpr, thresholds = roc_curve(y_true, y_score)
    fnr = 1 - tpr
    eer_idx = np.nanargmin(np.abs(fnr - fpr))
    eer = (fpr[eer_idx] + fnr[eer_idx]) / 2
    return eer, fpr, fnr, thresholds

# Plotting functions remain the same
def plot_det_curve(eers_data: dict, title: str):
    plt.figure(figsize=(8, 8))
    for label, (eer, fpr, fnr) in eers_data.items():
        if fpr is not None and fnr is not None:
            plt.plot(fpr * 100, fnr * 100, label=f"{label} (EER: {eer*100:.2f}%)")
    plt.title(title)
    plt.xlabel("False Positive Rate (%)")
    plt.ylabel("False Negative Rate (%)")
    plt.legend()
    plt.grid(True, which='both', linestyle='--')
    logging.info(f"Displaying DET plot: {title}")
    plt.show()

def evaluate_and_plot_confusion_matrix(gmms: dict, features_test: dict, verbose: bool = True):
    speakers = list(gmms.keys())
    y_true_labels, y_pred_labels = [], []

    iterator = tqdm(speakers, desc="Evaluating Identification on test clips") if verbose else speakers
    for true_spk in iterator:
        if true_spk not in features_test: continue
        for clip_feats in features_test[true_spk]:
            log_likelihoods = {pred_spk: gmms[pred_spk].score(clip_feats) for pred_spk in speakers if pred_spk in gmms}
            if not log_likelihoods: continue
            pred_spk = max(log_likelihoods, key=log_likelihoods.get)
            y_true_labels.append(true_spk)
            y_pred_labels.append(pred_spk)

    if not y_true_labels:
        logging.error("Could not generate predictions for confusion matrix.")
        return
    cm = confusion_matrix(y_true_labels, y_pred_labels, labels=speakers, normalize='true')
    disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=speakers)
    fig, ax = plt.subplots(figsize=(10, 10))
    disp.plot(ax=ax, xticks_rotation='vertical', cmap='viridis')
    plt.title("Speaker Identification Confusion Matrix (on Test Set)")
    plt.tight_layout()
    logging.info("Displaying confusion matrix.")
    plt.show()

# ================================
# 5. Main Experiment Runner
# ================================
def main(args):
    logging.info("="*30)
    logging.info("Experiment Configuration:")
    # ... print args ...
    logging.info(f"  Data Root: {args.data_root}, Test Size: {args.test_size}, MFCCs: {args.n_mfcc}")
    logging.info(f"  Deltas: {not args.no_deltas}, VAD: {not args.no_vad}, Mix Components: {args.n_mix_list}")
    logging.info("="*30)

    feature_params = {'n_mfcc': args.n_mfcc, 'use_deltas': not args.no_deltas, 'use_vad': not args.no_vad}
    train_data, test_data = load_data_and_split(args.data_root, test_size=args.test_size, **feature_params)
    if not train_data or not test_data:
        logging.error("Data loading/splitting failed. Exiting.")
        return
    logging.info(f"Data split complete. Training on {len(train_data)}, testing on {len(test_data)} speakers.")

    # --- Experiment (i): Conclusive Effect of Unbalanced Dataset ---
    logging.info("\n(i) Testing conclusive effect of unbalanced dataset")
    imbalance_levels = [0.0, 0.25, 0.50, 0.75]
    unbalanced_results = {}
    speakers_to_reduce = random.sample(list(train_data.keys()), k=len(train_data) // 2)
    logging.info(f"Will reduce data for speakers: {speakers_to_reduce}")

    for level in imbalance_levels:
        logging.info(f"Creating training set with {level*100}% data reduction.")
        unbalanced_train_data = {}
        for spk, feats in train_data.items():
            if spk in speakers_to_reduce and level > 0:
                reduce_to_frames = int(feats.shape[0] * (1 - level))
                if reduce_to_frames > 0:
                    unbalanced_train_data[spk] = feats[:reduce_to_frames, :]
            else:
                unbalanced_train_data[spk] = feats

        gmms = train_gmms(unbalanced_train_data, args.n_mix_unbalanced, verbose=False)
        eer, fpr, fnr, _ = evaluate_eer_per_clip(gmms, test_data, verbose=False) # USE THE NEW FUNCTION
        unbalanced_results[f"{int(level*100)}% reduction"] = (eer, fpr, fnr)
        logging.info(f"Imbalance Level {level*100}% -> EER = {eer*100:.2f}%")

    plt.figure(figsize=(10, 7))
    eers = [res[0] * 100 for res in unbalanced_results.values()]
    plt.plot(imbalance_levels, eers, marker='o')
    plt.title("Effect of Data Imbalance on EER (Per-Clip Evaluation)")
    plt.xlabel("Percentage of Data Removed for Half of Speakers")
    plt.ylabel("Equal Error Rate (EER) [%]")
    plt.xticks(imbalance_levels, [f'{int(l*100)}%' for l in imbalance_levels])
    plt.grid(True)
    plt.show()
    plot_det_curve(unbalanced_results, "DET Curve: Effect of Data Imbalance")

    # --- Experiment (ii): Effect of Number of Mixture Components ---
    logging.info("\n(ii) Effect of number of mixture components")
    n_components_list = [int(n) for n in args.n_mix_list.split(',')]
    eers_mix = []
    for n_mix in n_components_list:
        gmms = train_gmms(train_data, n_mix)
        eer, _, _, _ = evaluate_eer_per_clip(gmms, test_data) # USE THE NEW FUNCTION
        eers_mix.append(eer)
        logging.info(f"n_mix={n_mix}, EER={eer*100:.2f}%")
    plt.figure()
    plt.plot(n_components_list, [e*100 for e in eers_mix], marker='o')
    plt.title("Effect of Mixture Components (Evaluated on Test Set)")
    plt.xlabel("Number of GMM Components")
    plt.ylabel("EER (%)")
    plt.grid(True)
    plt.show()

    # --- Final Evaluation: Confusion Matrix ---
    logging.info("\n(iv) Generating final confusion matrix on test set")
    gmms_final = train_gmms(train_data, args.n_mix_unbalanced)
    if gmms_final:
        evaluate_and_plot_confusion_matrix(gmms_final, test_data)

    logging.info("\nâœ… Experiment completed successfully!")


if __name__ == "__main__":
    parser = argparse.ArgumentParser(description="Advanced GMM-based Speaker Recognition Experiment")
    parser.add_argument("--test_size", type=float, default=0.2, help="Proportion of the dataset for testing.")
    parser.add_argument("--data_root", type=str, default="data/data_clips", help="Root directory of speaker data.")
    parser.add_argument("--n_mfcc", type=int, default=20, help="Number of MFCCs.")
    parser.add_argument("--no_deltas", action='store_true', help="Disable delta features.")
    parser.add_argument("--no_vad", action='store_true', help="Disable VAD.")
    parser.add_argument("--n_mix_unbalanced", type=int, default=8, help="Mixture components for imbalance experiment.")
    parser.add_argument("--n_mix_list", type=str, default="2,4,8,16", help="Comma-separated list of mixture components.")

    args, unknown = parser.parse_known_args()
    main(args)

```

```python
import os
import numpy as np
import librosa
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.mixture import GaussianMixture
from sklearn.metrics import roc_curve, confusion_matrix, ConfusionMatrixDisplay
from sklearn.model_selection import train_test_split
from tqdm import tqdm
import argparse
import logging
import warnings
import random
import time

# Suppress warnings from librosa
warnings.filterwarnings('ignore', category=UserWarning, module='librosa')

# ================================
# Setup Logging
# ================================
logging.basicConfig(level=logging.INFO,
                    format='%(asctime)s - %(levelname)s - %(message)s')

# ================================
# 1. Advanced Data Loading & Feature Extraction with Train/Test Split
# ================================
def extract_features(file_path: str, n_mfcc: int = 20, use_deltas: bool = True, use_vad: bool = True) -> np.ndarray:
    """Extracts advanced features from a single audio file."""
    try:
        y, sr = librosa.load(file_path, sr=None)
        if use_vad:
            non_silent_intervals = librosa.effects.split(y, top_db=25)
            if non_silent_intervals.size == 0:
                logging.warning(f"No speech detected in {os.path.basename(file_path)}")
                return np.array([])
            y = np.concatenate([y[start:end] for start, end in non_silent_intervals])
        if len(y) == 0: return np.array([])
        mfccs = librosa.feature.mfcc(y=y, sr=sr, n_mfcc=n_mfcc)
        mfccs = (mfccs - np.mean(mfccs, axis=1, keepdims=True)) / (np.std(mfccs, axis=1, keepdims=True) + 1e-8)
        if use_deltas:
            if mfccs.shape[1] < 9:
                logging.warning(f"File {os.path.basename(file_path)} is too short for deltas. Skipping.")
                return np.array([])
            delta1 = librosa.feature.delta(mfccs, order=1)
            delta2 = librosa.feature.delta(mfccs, order=2)
            features = np.vstack([mfccs, delta1, delta2])
        else:
            features = mfccs
        return features.T
    except Exception as e:
        logging.error(f"Error processing {file_path}: {e}")
        return np.array([])

def get_train_test_files(data_root: str, test_size: float = 0.2):
    """Splits audio files for each speaker into train and test sets."""
    train_files_dict, test_files_dict = {}, {}
    speaker_paths = [os.path.join(data_root, spk) for spk in sorted(os.listdir(data_root))]
    speaker_paths = [p for p in speaker_paths if os.path.isdir(p)]

    for spk_path in speaker_paths:
        spk = os.path.basename(spk_path)
        files = [os.path.join(spk_path, f) for f in os.listdir(spk_path) if f.lower().endswith('.wav')]
        if len(files) < 2:
            logging.warning(f"Speaker {spk} has < 2 clips, cannot split. Skipping.")
            continue
        train_files, test_files = train_test_split(files, test_size=test_size, random_state=42)
        train_files_dict[spk] = train_files
        test_files_dict[spk] = test_files

    return train_files_dict, test_files_dict

def process_files_to_feature_dict(files_dict: dict, concatenate: bool, verbose: bool = True, **feature_params):
    """Processes file lists into feature dictionaries."""
    features_dict = {}

    iterator = tqdm(files_dict.items(), desc="Extracting features") if verbose else files_dict.items()

    for spk, files in iterator:
        feats_list = [extract_features(f, **feature_params) for f in files]
        feats_list = [f for f in feats_list if f.shape[0] > 0]

        if not feats_list:
            continue

        if concatenate:
            features_dict[spk] = np.vstack(feats_list)
        else:
            features_dict[spk] = feats_list

    return features_dict


# ================================
# 2. Train GMMs
# ================================
def train_gmms(features: dict, n_mix: int, verbose: bool = True) -> dict:
    gmms = {}
    iterator = tqdm(features.items(), desc=f"Training GMMs (n_mix={n_mix})") if verbose else features.items()
    for spk, feats in iterator:
        n_frames = feats.shape[0]
        effective_n_mix = n_mix
        if n_frames < n_mix * 10:
            effective_n_mix = max(1, n_frames // 10)
            if verbose: logging.warning(f"{spk}: Not enough data for {n_mix} components. Using {effective_n_mix}.")
        try:
            gmm = GaussianMixture(n_components=effective_n_mix, covariance_type='diag', reg_covar=1e-4, max_iter=200, n_init=3, random_state=0)
            gmm.fit(feats)
            gmms[spk] = gmm
        except ValueError as e:
            logging.error(f"Skipping {spk}: GMM fitting failed. {e}")
            continue
    return gmms

# ================================
# 3. Advanced Evaluation (per-clip)
# ================================
def evaluate_eer_per_clip(gmms: dict, features_test: dict, verbose: bool = True) -> tuple:
    speakers = list(gmms.keys())
    scores_true, scores_false = [], []
    iterator = tqdm(speakers, desc="Evaluating EER on test clips") if verbose else speakers
    for true_spk in iterator:
        if true_spk not in features_test or true_spk not in gmms: continue
        for clip_feats in features_test[true_spk]:
            scores_true.append(gmms[true_spk].score(clip_feats))
            for imposter_spk in speakers:
                if imposter_spk != true_spk and imposter_spk in gmms:
                    scores_false.append(gmms[imposter_spk].score(clip_feats))
    if not scores_true or not scores_false:
        logging.error("Evaluation failed: No scores generated.")
        return 1.0, None, None, None
    y_true = np.array([1] * len(scores_true) + [0] * len(scores_false))
    y_score = np.array(scores_true + scores_false)
    fpr, tpr, thresholds = roc_curve(y_true, y_score)
    fnr = 1 - tpr
    eer_idx = np.nanargmin(np.abs(fnr - fpr))
    eer = (fpr[eer_idx] + fnr[eer_idx]) / 2
    return eer, fpr, fnr, thresholds

def plot_det_curve(eers_data: dict, title: str):
    plt.figure(figsize=(8, 8))
    for label, (eer, fpr, fnr) in eers_data.items():
        if fpr is not None and fnr is not None:
            plt.plot(fpr * 100, fnr * 100, label=f"{label} (EER: {eer*100:.2f}%)")
    plt.title(title)
    plt.xlabel("False Positive Rate (%)")
    plt.ylabel("False Negative Rate (%)")
    plt.legend()
    plt.grid(True, which='both', linestyle='--')
    plt.show()

def evaluate_and_plot_confusion_matrix(gmms: dict, features_test: dict, verbose: bool = True):
    speakers = list(gmms.keys())
    y_true_labels, y_pred_labels = [], []
    iterator = tqdm(speakers, desc="Evaluating Identification on test clips") if verbose else speakers
    for true_spk in iterator:
        if true_spk not in features_test: continue
        for clip_feats in features_test[true_spk]:
            log_likelihoods = {pred_spk: gmms[pred_spk].score(clip_feats) for pred_spk in speakers if pred_spk in gmms}
            if not log_likelihoods: continue
            pred_spk = max(log_likelihoods, key=log_likelihoods.get)
            y_true_labels.append(true_spk)
            y_pred_labels.append(pred_spk)
    if not y_true_labels:
        logging.error("Could not generate predictions for confusion matrix.")
        return
    cm = confusion_matrix(y_true_labels, y_pred_labels, labels=speakers, normalize='true')
    disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=speakers)
    fig, ax = plt.subplots(figsize=(10, 10))
    disp.plot(ax=ax, xticks_rotation='vertical', cmap='viridis')
    plt.title("Speaker Identification Confusion Matrix (on Test Set)")
    plt.tight_layout()
    plt.show()

# ================================
# 5. Main Experiment Runner
# ================================
def main(args):
    logging.info("="*30 + "\nExperiment Configuration:")
    logging.info(f"  Data Root: {args.data_root}, Test Size: {args.test_size}, MFCCs: {args.n_mfcc}")
    logging.info(f"  Deltas: {not args.no_deltas}, VAD: {not args.no_vad}")
    logging.info(f"  Mix Components: {args.n_mix_list}")
    logging.info("="*30)

    feature_params = {'n_mfcc': args.n_mfcc, 'use_deltas': not args.no_deltas, 'use_vad': not args.no_vad}

    train_files_dict, test_files_dict = get_train_test_files(args.data_root, args.test_size)
    if not train_files_dict or not test_files_dict:
        logging.error("Could not find or split data. Exiting.")
        return

    logging.info("Processing test files...")
    test_data = process_files_to_feature_dict(test_files_dict, concatenate=False, **feature_params)
    logging.info("Processing full training files...")
    full_train_data = process_files_to_feature_dict(train_files_dict, concatenate=True, **feature_params)

    if not test_data or not full_train_data:
        logging.error("Feature extraction failed for train/test sets. Exiting.")
        return
    logging.info(f"Data split and processed. Training on {len(full_train_data)}, testing on {len(test_data)} speakers.")

    # --- Experiment (i): Conclusive Effect of Unbalanced Dataset ---
    logging.info("\n(i) Testing conclusive effect of unbalanced dataset")
    imbalance_levels = [0.0, 0.25, 0.50, 0.75]
    unbalanced_results = {}
    speakers_to_reduce = random.sample(list(full_train_data.keys()), k=len(full_train_data) // 2)
    logging.info(f"Will reduce data for speakers: {speakers_to_reduce}")

    for level in imbalance_levels:
        logging.info(f"Creating training set with {level*100}% data reduction.")
        unbalanced_train_data = {}
        for spk, feats in full_train_data.items():
            if spk in speakers_to_reduce and level > 0:
                reduce_to_frames = int(feats.shape[0] * (1 - level))
                if reduce_to_frames > 0:
                    unbalanced_train_data[spk] = feats[:reduce_to_frames, :]
            else:
                unbalanced_train_data[spk] = feats

        gmms = train_gmms(unbalanced_train_data, args.n_mix_unbalanced, verbose=False)
        eer, fpr, fnr, _ = evaluate_eer_per_clip(gmms, test_data, verbose=False)
        unbalanced_results[f"{int(level*100)}% reduction"] = (eer, fpr, fnr)
        logging.info(f"Imbalance Level {level*100}% -> EER = {eer*100:.2f}%")

    plt.figure(figsize=(10, 7))
    eers = [res[0] * 100 for res in unbalanced_results.values()]
    plt.plot(imbalance_levels, eers, marker='o')
    plt.title("Effect of Data Imbalance on EER (Per-Clip Evaluation)")
    plt.xlabel("Percentage of Data Removed for Half of Speakers")
    plt.ylabel("Equal Error Rate (EER) [%]")
    plt.xticks(imbalance_levels, [f'{int(l*100)}%' for l in imbalance_levels])
    plt.grid(True)
    plt.show()
    plot_det_curve(unbalanced_results, "DET Curve: Effect of Data Imbalance")

    # --- Experiment (ii): Effect of Number of Mixture Components & training time ---
    logging.info("\n(ii) Effect of number of mixture components & training time")
    n_components_list = [int(n) for n in args.n_mix_list.split(',')]
    eers_mix, training_times = [], []
    for n_mix in n_components_list:
        start_time = time.time()
        gmms = train_gmms(full_train_data, n_mix)
        duration = time.time() - start_time
        training_times.append(duration)
        eer, _, _, _ = evaluate_eer_per_clip(gmms, test_data)
        eers_mix.append(eer)
        logging.info(f"n_mix={n_mix}, EER={eer*100:.2f}%, Training Time={duration:.2f}s")

    # EER Plot
    plt.figure()
    plt.plot(n_components_list, [e*100 for e in eers_mix], marker='o')
    plt.title("Effect of Mixture Components (Evaluated on Test Set)")
    plt.xlabel("Number of GMM Components")
    plt.ylabel("EER (%)")
    plt.grid(True)
    plt.show()

    # Training Time Plot
    plt.figure()
    plt.plot(n_components_list, training_times, marker='o', color='r')
    plt.title("Training Time vs. Number of Mixture Components")
    plt.xlabel("Number of GMM Components")
    plt.ylabel("Training Time (seconds)")
    plt.grid(True)
    plt.show()

    # --- Final Evaluation: Confusion Matrix ---
    logging.info("\n(iv) Generating final confusion matrix on test set")
    gmms_final = train_gmms(full_train_data, args.n_mix_unbalanced)
    if gmms_final:
        evaluate_and_plot_confusion_matrix(gmms_final, test_data)

    logging.info("\nâœ… Experiment completed successfully!")


if __name__ == "__main__":
    parser = argparse.ArgumentParser(description="Advanced GMM-based Speaker Recognition Experiment")
    parser.add_argument("--test_size", type=float, default=0.2, help="Proportion of the dataset for testing.")
    parser.add_argument("--data_root", type=str, default="data/data_clips", help="Root directory of speaker data.")
    parser.add_argument("--n_mfcc", type=int, default=20, help="Number of MFCCs.")
    parser.add_argument("--no_deltas", action='store_true', help="Disable delta features.")
    parser.add_argument("--no_vad", action='store_true', help="Disable VAD.")
    parser.add_argument("--n_mix_unbalanced", type=int, default=8, help="Mixture components for imbalance experiment.")
    parser.add_argument("--n_mix_list", type=str, default="2,4,8,16", help="Comma-separated list of mixture components.")

    args, unknown = parser.parse_known_args()
    main(args)

```

```python
import os
import numpy as np
import librosa
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.mixture import GaussianMixture
from sklearn.metrics import roc_curve, confusion_matrix, ConfusionMatrixDisplay
from sklearn.model_selection import train_test_split
from tqdm import tqdm
import argparse
import warnings
import random
import time

# Suppress warnings from librosa
warnings.filterwarnings('ignore', category=UserWarning, module='librosa')

# ================================
# 1. Advanced Data Loading & Feature Extraction with Train/Test Split
# ================================
def extract_features(file_path: str, n_mfcc: int, n_fft: int, hop_length: int, n_mels: int, use_deltas: bool = True, use_vad: bool = True) -> np.ndarray:
    """Extracts advanced features from a single audio file."""
    try:
        y, sr = librosa.load(file_path, sr=None)
        if use_vad:
            non_silent_intervals = librosa.effects.split(y, top_db=25)
            if non_silent_intervals.size == 0:
                print(f"WARNING: No speech detected in {os.path.basename(file_path)}")
                return np.array([])
            y = np.concatenate([y[start:end] for start, end in non_silent_intervals])
        if len(y) == 0: return np.array([])
        mfccs = librosa.feature.mfcc(y=y, sr=sr, n_mfcc=n_mfcc, n_fft=n_fft, hop_length=hop_length, n_mels=n_mels)
        mfccs = (mfccs - np.mean(mfccs, axis=1, keepdims=True)) / (np.std(mfccs, axis=1, keepdims=True) + 1e-8)
        if use_deltas:
            if mfccs.shape[1] < 9:
                print(f"WARNING: File {os.path.basename(file_path)} is too short for deltas. Skipping.")
                return np.array([])
            delta1 = librosa.feature.delta(mfccs, order=1)
            delta2 = librosa.feature.delta(mfccs, order=2)
            features = np.vstack([mfccs, delta1, delta2])
        else:
            features = mfccs
        return features.T
    except Exception as e:
        print(f"ERROR: Error processing {file_path}: {e}")
        return np.array([])

def get_train_test_files(data_root: str, test_size: float = 0.2):
    """Splits audio files for each speaker into train and test sets."""
    train_files_dict, test_files_dict = {}, {}
    speaker_paths = [os.path.join(data_root, spk) for spk in sorted(os.listdir(data_root))]
    speaker_paths = [p for p in speaker_paths if os.path.isdir(p)]

    for spk_path in speaker_paths:
        spk = os.path.basename(spk_path)
        files = [os.path.join(spk_path, f) for f in os.listdir(spk_path) if f.lower().endswith('.wav')]
        if len(files) < 2:
            print(f"WARNING: Speaker {spk} has < 2 clips, cannot split. Skipping.")
            continue
        train_files, test_files = train_test_split(files, test_size=test_size, random_state=42)
        train_files_dict[spk] = train_files
        test_files_dict[spk] = test_files

    return train_files_dict, test_files_dict

def process_files_to_feature_dict(files_dict: dict, concatenate: bool, verbose: bool = True, **feature_params):
    """Processes file lists into feature dictionaries."""
    features_dict = {}

    iterator = tqdm(files_dict.items(), desc="Extracting features") if verbose else files_dict.items()

    for spk, files in iterator:
        feats_list = [extract_features(f, **feature_params) for f in files]
        feats_list = [f for f in feats_list if f.shape[0] > 0]

        if not feats_list:
            continue

        if concatenate:
            features_dict[spk] = np.vstack(feats_list)
        else:
            features_dict[spk] = feats_list

    return features_dict


# ================================
# 2. Train GMMs
# ================================
def train_gmms(features: dict, n_mix: int, covariance_type: str, max_iter: int, n_init: int, verbose: bool = True) -> dict:
    gmms = {}
    iterator = tqdm(features.items(), desc=f"Training GMMs (n_mix={n_mix})") if verbose else features.items()
    for spk, feats in iterator:
        n_frames = feats.shape[0]
        effective_n_mix = n_mix
        if n_frames < n_mix * 10:
            effective_n_mix = max(1, n_frames // 10)
            if verbose: print(f"WARNING: {spk}: Not enough data for {n_mix} components. Using {effective_n_mix}.")
        try:
            gmm = GaussianMixture(n_components=effective_n_mix,
                                  covariance_type=covariance_type,
                                  reg_covar=1e-4,
                                  max_iter=max_iter,
                                  n_init=n_init,
                                  random_state=0)
            gmm.fit(feats)
            gmms[spk] = gmm
        except ValueError as e:
            print(f"ERROR: Skipping {spk}: GMM fitting failed. {e}")
            continue
    return gmms

# ================================
# 3. Advanced Evaluation (per-clip)
# ================================
def evaluate_eer_per_clip(gmms: dict, features_test: dict, verbose: bool = True) -> tuple:
    speakers = list(gmms.keys())
    scores_true, scores_false = [], []
    iterator = tqdm(speakers, desc="Evaluating EER on test clips") if verbose else speakers
    for true_spk in iterator:
        if true_spk not in features_test or true_spk not in gmms: continue
        for clip_feats in features_test[true_spk]:
            scores_true.append(gmms[true_spk].score(clip_feats))
            for imposter_spk in speakers:
                if imposter_spk != true_spk and imposter_spk in gmms:
                    scores_false.append(gmms[imposter_spk].score(clip_feats))
    if not scores_true or not scores_false:
        print("ERROR: Evaluation failed: No scores generated.")
        return 1.0, None, None, None
    y_true = np.array([1] * len(scores_true) + [0] * len(scores_false))
    y_score = np.array(scores_true + scores_false)
    fpr, tpr, thresholds = roc_curve(y_true, y_score)
    fnr = 1 - tpr
    eer_idx = np.nanargmin(np.abs(fnr - fpr))
    eer = (fpr[eer_idx] + fnr[eer_idx]) / 2
    return eer, fpr, fnr, thresholds

def plot_det_curve(eers_data: dict, title: str):
    plt.figure(figsize=(8, 8))
    for label, (eer, fpr, fnr) in eers_data.items():
        if fpr is not None and fnr is not None:
            plt.plot(fpr * 100, fnr * 100, label=f"{label} (EER: {eer*100:.2f}%)")
    plt.title(title)
    plt.xlabel("False Positive Rate (%)")
    plt.ylabel("False Negative Rate (%)")
    plt.legend()
    plt.grid(True, which='both', linestyle='--')
    plt.show()

def evaluate_and_plot_confusion_matrix(gmms: dict, features_test: dict, verbose: bool = True):
    speakers = list(gmms.keys())
    y_true_labels, y_pred_labels = [], []
    iterator = tqdm(speakers, desc="Evaluating Identification on test clips") if verbose else speakers
    for true_spk in iterator:
        if true_spk not in features_test: continue
        for clip_feats in features_test[true_spk]:
            log_likelihoods = {pred_spk: gmms[pred_spk].score(clip_feats) for pred_spk in speakers if pred_spk in gmms}
            if not log_likelihoods: continue
            pred_spk = max(log_likelihoods, key=log_likelihoods.get)
            y_true_labels.append(true_spk)
            y_pred_labels.append(pred_spk)
    if not y_true_labels:
        print("ERROR: Could not generate predictions for confusion matrix.")
        return
    cm = confusion_matrix(y_true_labels, y_pred_labels, labels=speakers, normalize='true')
    disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=speakers)
    fig, ax = plt.subplots(figsize=(10, 10))
    disp.plot(ax=ax, xticks_rotation='vertical', cmap='viridis')
    plt.title("Speaker Identification Confusion Matrix (on Test Set)")
    plt.tight_layout()
    plt.show()

# ================================
# 5. Main Experiment Runner
# ================================
def main(args):
    print("="*30 + "\nExperiment Configuration:")
    print(f"  Data Root: {args.data_root}, Test Size: {args.test_size}")
    print("-" * 20)
    print("  Feature Extraction Parameters:")
    print(f"    Number of MFCCs: {args.n_mfcc}")
    print(f"    FFT Window Size: {args.n_fft}")
    print(f"    Hop Length: {args.hop_length}")
    print(f"    Number of Mel Bands: {args.n_mels}")
    print(f"    Use Deltas: {not args.no_deltas}")
    print(f"    Use VAD: {not args.no_vad}")
    print("-" * 20)
    print("  GMM Training Parameters:")
    print(f"    Mixture Components: {args.n_mix_list}")
    print(f"    Covariance Type: {args.covariance_type}")
    print(f"    Max EM Iterations: {args.max_iter}")
    print(f"    Number of Initializations: {args.n_init}")
    print("="*30)

    feature_params = {
        'n_mfcc': args.n_mfcc, 'n_fft': args.n_fft, 'hop_length': args.hop_length,
        'n_mels': args.n_mels, 'use_deltas': not args.no_deltas, 'use_vad': not args.no_vad
    }

    gmm_params = {
        'covariance_type': args.covariance_type, 'max_iter': args.max_iter, 'n_init': args.n_init
    }

    train_files_dict, test_files_dict = get_train_test_files(args.data_root, args.test_size)
    if not train_files_dict or not test_files_dict:
        print("ERROR: Could not find or split data. Exiting.")
        return

    print("Processing test files...")
    test_data = process_files_to_feature_dict(test_files_dict, concatenate=False, **feature_params)
    print("Processing full training files...")
    full_train_data = process_files_to_feature_dict(train_files_dict, concatenate=True, **feature_params)

    if not test_data or not full_train_data:
        print("ERROR: Feature extraction failed for train/test sets. Exiting.")
        return
    print(f"Data split and processed. Training on {len(full_train_data)}, testing on {len(test_data)} speakers.")

    # --- Experiment (i): Conclusive Effect of Unbalanced Dataset ---
    print("\n(i) Testing conclusive effect of unbalanced dataset")
    imbalance_levels = [0.0, 0.25, 0.50, 0.75]
    unbalanced_results = {}
    speakers_to_reduce = random.sample(list(full_train_data.keys()), k=len(full_train_data) // 2)
    print(f"Will reduce data for speakers: {speakers_to_reduce}")

    for level in imbalance_levels:
        unbalanced_train_data = {}
        for spk, feats in full_train_data.items():
            if spk in speakers_to_reduce and level > 0:
                reduce_to_frames = int(feats.shape[0] * (1 - level))
                if reduce_to_frames > 0:
                    unbalanced_train_data[spk] = feats[:reduce_to_frames, :]
            else:
                unbalanced_train_data[spk] = feats

        gmms = train_gmms(unbalanced_train_data, args.n_mix_unbalanced, verbose=False, **gmm_params)
        eer, fpr, fnr, _ = evaluate_eer_per_clip(gmms, test_data, verbose=False)
        unbalanced_results[f"{int(level*100)}% reduction"] = (eer, fpr, fnr)
        print(f"Imbalance Level {level*100}% -> EER = {eer*100:.2f}%")

    # Plotting for imbalance experiment
    plt.figure(figsize=(10, 7))
    eers_unbalanced = [res[0] * 100 for res in unbalanced_results.values()]
    plt.plot(imbalance_levels, eers_unbalanced, marker='o')
    plt.title("Effect of Data Imbalance on EER (Per-Clip Evaluation)")
    plt.xlabel("Percentage of Data Removed for Half of Speakers")
    plt.ylabel("Equal Error Rate (EER) [%]")
    plt.xticks(imbalance_levels, [f'{int(l*100)}%' for l in imbalance_levels])
    plt.grid(True)
    plt.show()
    plot_det_curve(unbalanced_results, "DET Curve: Effect of Data Imbalance")

    # --- Experiment (ii): Effect of Number of Mixture Components & training time ---
    print("\n(ii) Effect of number of mixture components & training time")
    n_components_list = [int(n) for n in args.n_mix_list.split(',')]
    eers_mix, training_times = [], []
    for n_mix in n_components_list:
        start_time = time.time()
        gmms = train_gmms(full_train_data, n_mix, **gmm_params)
        duration = time.time() - start_time
        training_times.append(duration)
        eer, _, _, _ = evaluate_eer_per_clip(gmms, test_data)
        eers_mix.append(eer)
        print(f"n_mix={n_mix}, EER={eer*100:.2f}%, Training Time={duration:.2f}s")

    # Plotting for EER and Training time
    fig, ax1 = plt.subplots(figsize=(10, 7))
    ax1.set_xlabel('Number of GMM Components')
    ax1.set_ylabel('Equal Error Rate (EER) [%]', color='b')
    ax1.plot(n_components_list, [e * 100 for e in eers_mix], marker='o', color='b', label='EER')
    ax1.tick_params(axis='y', labelcolor='b')

    ax2 = ax1.twinx()
    ax2.set_ylabel('Training Time (seconds)', color='r')
    ax2.plot(n_components_list, training_times, marker='o', color='r', label='Training Time')
    ax2.tick_params(axis='y', labelcolor='r')

    fig.tight_layout()
    plt.title('EER and Training Time vs. Number of GMM Components')
    plt.grid(True)
    plt.show()

    # --- Experiment (iii): Amount of Data vs. Number of Mixture Components Heatmap ---
    print("\n(iii) Generating EER heatmap for Data Amount vs. GMM Components")
    clip_counts = [int(c) for c in args.clip_counts.split(',')]
    heatmap_eers = np.zeros((len(clip_counts), len(n_components_list)))

    for i, num_clips in enumerate(clip_counts):
        print(f"--- Running heatmap experiment for {num_clips} training clips per speaker ---")
        # Ensure enough speakers have the required number of clips
        valid_speakers = {spk: files for spk, files in train_files_dict.items() if len(files) >= num_clips}
        if len(valid_speakers) < 2: # Need at least 2 speakers to evaluate
            print(f"WARNING: Not enough speakers with at least {num_clips} clips. Skipping.")
            heatmap_eers[i, :] = np.nan # Mark row as invalid
            continue

        subset_train_files = {spk: files[:num_clips] for spk, files in valid_speakers.items()}
        subset_train_data = process_files_to_feature_dict(subset_train_files, concatenate=True, verbose=False, **feature_params)

        for j, n_mix in enumerate(n_components_list):
            gmms = train_gmms(subset_train_data, n_mix, verbose=False, **gmm_params)
            # Ensure we only evaluate on speakers that were part of training
            test_data_subset = {spk: clips for spk, clips in test_data.items() if spk in gmms}
            if len(test_data_subset) < 2:
                eer = 1.0 # Max error if not enough speakers to test
            else:
                eer, _, _, _ = evaluate_eer_per_clip(gmms, test_data_subset, verbose=False)
            heatmap_eers[i, j] = eer * 100
            print(f"  clips={num_clips}, components={n_mix}, EER={eer*100:.2f}%")

    plt.figure(figsize=(12, 8))
    sns.heatmap(heatmap_eers, annot=True, fmt=".2f", cmap="viridis_r",
                xticklabels=n_components_list, yticklabels=clip_counts)
    plt.xlabel("Number of GMM Components")
    plt.ylabel("Number of Training Clips per Speaker")
    plt.title("EER (%) Heatmap: GMM Components vs. Training Data Amount")
    plt.show()

    # --- Final Evaluation: Confusion Matrix ---
    print("\n(iv) Generating final confusion matrix on test set")
    gmms_final = train_gmms(full_train_data, args.n_mix_unbalanced, **gmm_params)
    if gmms_final:
        evaluate_and_plot_confusion_matrix(gmms_final, test_data)

    print("\nâœ… Experiment completed successfully!")


if __name__ == "__main__":
    parser = argparse.ArgumentParser(description="Advanced GMM-based Speaker Recognition Experiment")

    # General arguments
    parser.add_argument("--data_root", type=str, default="data/data_clips", help="Root directory of speaker data.")
    parser.add_argument("--test_size", type=float, default=0.2, help="Proportion of the dataset for testing.")

    # Feature extraction arguments
    parser.add_argument("--n_mfcc", type=int, default=20, help="Number of MFCCs to extract.")
    parser.add_argument("--n_fft", type=int, default=2048, help="FFT window size for MFCC.")
    parser.add_argument("--hop_length", type=int, default=512, help="Hop length for MFCC.")
    parser.add_argument("--n_mels", type=int, default=128, help="Number of Mel bands for MFCC.")
    parser.add_argument("--no_deltas", action='store_true', help="Disable delta features.")
    parser.add_argument("--no_vad", action='store_true', help="Disable VAD.")

    # GMM training arguments
    parser.add_argument("--n_mix_list", type=str, default="2,4,8,16", help="Comma-separated list of mixture components to test.")
    parser.add_argument("--n_mix_unbalanced", type=int, default=8, help="Mixture components for imbalance experiment.")
    parser.add_argument("--covariance_type", type=str, default="diag", choices=['full', 'tied', 'diag', 'spherical'], help="Covariance type for GMMs.")
    parser.add_argument("--max_iter", type=int, default=200, help="Max EM iterations for GMM training.")
    parser.add_argument("--n_init", type=int, default=3, help="Number of initializations for GMM training.")

    # Heatmap specific argument
    parser.add_argument("--clip_counts", type=str, default="2,4,8,12", help="Comma-separated list of clip counts per speaker for heatmap.")

    args, unknown = parser.parse_known_args()
    main(args)

```

```python
import os
import numpy as np
import librosa
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.mixture import GaussianMixture
from sklearn.metrics import roc_curve, confusion_matrix, ConfusionMatrixDisplay
from sklearn.model_selection import train_test_split
from tqdm import tqdm
import argparse
import warnings
import random
import time

# Suppress warnings from librosa
warnings.filterwarnings('ignore', category=UserWarning, module='librosa')

# ================================
# 1. Advanced Data Loading & Feature Extraction with Train/Test Split
# ================================
def extract_features(file_path: str, n_mfcc: int, n_fft: int, hop_length: int, n_mels: int, use_deltas: bool = True, use_vad: bool = True) -> np.ndarray:
    """Extracts advanced features from a single audio file."""
    try:
        y, sr = librosa.load(file_path, sr=None)
        if use_vad:
            non_silent_intervals = librosa.effects.split(y, top_db=25)
            if non_silent_intervals.size == 0:
                print(f"WARNING: No speech detected in {os.path.basename(file_path)}")
                return np.array([])
            y = np.concatenate([y[start:end] for start, end in non_silent_intervals])
        if len(y) == 0: return np.array([])
        mfccs = librosa.feature.mfcc(y=y, sr=sr, n_mfcc=n_mfcc, n_fft=n_fft, hop_length=hop_length, n_mels=n_mels)
        mfccs = (mfccs - np.mean(mfccs, axis=1, keepdims=True)) / (np.std(mfccs, axis=1, keepdims=True) + 1e-8)
        if use_deltas:
            if mfccs.shape[1] < 9:
                print(f"WARNING: File {os.path.basename(file_path)} is too short for deltas. Skipping.")
                return np.array([])
            delta1 = librosa.feature.delta(mfccs, order=1)
            delta2 = librosa.feature.delta(mfccs, order=2)
            features = np.vstack([mfccs, delta1, delta2])
        else:
            features = mfccs
        return features.T
    except Exception as e:
        print(f"ERROR: Error processing {file_path}: {e}")
        return np.array([])

def get_train_test_files(data_root: str, test_size: float = 0.2):
    """Splits audio files for each speaker into train and test sets."""
    train_files_dict, test_files_dict = {}, {}
    speaker_paths = [os.path.join(data_root, spk) for spk in sorted(os.listdir(data_root))]
    speaker_paths = [p for p in speaker_paths if os.path.isdir(p)]

    for spk_path in speaker_paths:
        spk = os.path.basename(spk_path)
        files = [os.path.join(spk_path, f) for f in os.listdir(spk_path) if f.lower().endswith('.wav')]
        if len(files) < 2:
            print(f"WARNING: Speaker {spk} has < 2 clips, cannot split. Skipping.")
            continue
        train_files, test_files = train_test_split(files, test_size=test_size, random_state=42)
        train_files_dict[spk] = train_files
        test_files_dict[spk] = test_files

    return train_files_dict, test_files_dict

def process_files_to_feature_dict(files_dict: dict, concatenate: bool, verbose: bool = True, **feature_params):
    """Processes file lists into feature dictionaries."""
    features_dict = {}

    iterator = tqdm(files_dict.items(), desc="Extracting features") if verbose else files_dict.items()

    for spk, files in iterator:
        feats_list = [extract_features(f, **feature_params) for f in files]
        feats_list = [f for f in feats_list if f.shape[0] > 0]

        if not feats_list:
            continue

        if concatenate:
            features_dict[spk] = np.vstack(feats_list)
        else:
            features_dict[spk] = feats_list

    return features_dict


# ================================
# 2. Train GMMs
# ================================
def train_gmms(features: dict, n_mix: int, covariance_type: str, max_iter: int, n_init: int, verbose: bool = True) -> dict:
    gmms = {}
    iterator = tqdm(features.items(), desc=f"Training GMMs (n_mix={n_mix})") if verbose else features.items()
    for spk, feats in iterator:
        n_frames = feats.shape[0]
        effective_n_mix = n_mix
        if n_frames < n_mix * 10:
            effective_n_mix = max(1, n_frames // 10)
            if verbose: print(f"WARNING: {spk}: Not enough data for {n_mix} components. Using {effective_n_mix}.")
        try:
            gmm = GaussianMixture(n_components=effective_n_mix,
                                  covariance_type=covariance_type,
                                  reg_covar=1e-4,
                                  max_iter=max_iter,
                                  n_init=n_init,
                                  random_state=0)
            gmm.fit(feats)
            gmms[spk] = gmm
        except ValueError as e:
            print(f"ERROR: Skipping {spk}: GMM fitting failed. {e}")
            continue
    return gmms

# ================================
# 3. Advanced Evaluation (per-clip)
# ================================
def evaluate_eer_per_clip(gmms: dict, features_test: dict, verbose: bool = True) -> tuple:
    speakers = list(gmms.keys())
    scores_true, scores_false = [], []
    iterator = tqdm(speakers, desc="Evaluating EER on test clips") if verbose else speakers
    for true_spk in iterator:
        if true_spk not in features_test or true_spk not in gmms: continue
        for clip_feats in features_test[true_spk]:
            scores_true.append(gmms[true_spk].score(clip_feats))
            for imposter_spk in speakers:
                if imposter_spk != true_spk and imposter_spk in gmms:
                    scores_false.append(gmms[imposter_spk].score(clip_feats))
    if not scores_true or not scores_false:
        print("ERROR: Evaluation failed: No scores generated.")
        return 1.0, None, None, None
    y_true = np.array([1] * len(scores_true) + [0] * len(scores_false))
    y_score = np.array(scores_true + scores_false)
    fpr, tpr, thresholds = roc_curve(y_true, y_score)
    fnr = 1 - tpr
    eer_idx = np.nanargmin(np.abs(fnr - fpr))
    eer = (fpr[eer_idx] + fnr[eer_idx]) / 2
    return eer, fpr, fnr, thresholds

def plot_det_curve(eers_data: dict, title: str):
    plt.figure(figsize=(8, 8))
    for label, (eer, fpr, fnr) in eers_data.items():
        if fpr is not None and fnr is not None:
            plt.plot(fpr * 100, fnr * 100, label=f"{label} (EER: {eer*100:.2f}%)")
    plt.title(title)
    plt.xlabel("False Positive Rate (%)")
    plt.ylabel("False Negative Rate (%)")
    plt.legend()
    plt.grid(True, which='both', linestyle='--')
    plt.show()

def evaluate_and_plot_confusion_matrix(gmms: dict, features_test: dict, verbose: bool = True):
    speakers = list(gmms.keys())
    y_true_labels, y_pred_labels = [], []
    iterator = tqdm(speakers, desc="Evaluating Identification on test clips") if verbose else speakers
    for true_spk in iterator:
        if true_spk not in features_test: continue
        for clip_feats in features_test[true_spk]:
            log_likelihoods = {pred_spk: gmms[pred_spk].score(clip_feats) for pred_spk in speakers if pred_spk in gmms}
            if not log_likelihoods: continue
            pred_spk = max(log_likelihoods, key=log_likelihoods.get)
            y_true_labels.append(true_spk)
            y_pred_labels.append(pred_spk)
    if not y_true_labels:
        print("ERROR: Could not generate predictions for confusion matrix.")
        return
    cm = confusion_matrix(y_true_labels, y_pred_labels, labels=speakers, normalize='true')
    disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=speakers)
    fig, ax = plt.subplots(figsize=(10, 10))
    disp.plot(ax=ax, xticks_rotation='vertical', cmap='viridis')
    plt.title("Speaker Identification Confusion Matrix (on Test Set)")
    plt.tight_layout()
    plt.show()

# ================================
# 5. Main Experiment Runner
# ================================
def main(args):
    print("="*30 + "\nExperiment Configuration:")
    print(f"  Data Root: {args.data_root}, Test Size: {args.test_size}")
    print("-" * 20)
    print("  Feature Extraction Parameters:")
    print(f"    Number of MFCCs: {args.n_mfcc}")
    print(f"    FFT Window Size: {args.n_fft}")
    print(f"    Hop Length: {args.hop_length}")
    print(f"    Number of Mel Bands: {args.n_mels}")
    print(f"    Use Deltas: {not args.no_deltas}")
    print(f"    Use VAD: {not args.no_vad}")
    print("-" * 20)
    print("  GMM Training Parameters:")
    print(f"    Mixture Components: {args.n_mix_list}")
    print(f"    Covariance Type: {args.covariance_type}")
    print(f"    Max EM Iterations: {args.max_iter}")
    print(f"    Number of Initializations: {args.n_init}")
    print("="*30)

    feature_params = {
        'n_mfcc': args.n_mfcc, 'n_fft': args.n_fft, 'hop_length': args.hop_length,
        'n_mels': args.n_mels, 'use_deltas': not args.no_deltas, 'use_vad': not args.no_vad
    }

    gmm_params = {
        'covariance_type': args.covariance_type, 'max_iter': args.max_iter, 'n_init': args.n_init
    }

    train_files_dict, test_files_dict = get_train_test_files(args.data_root, args.test_size)
    if not train_files_dict or not test_files_dict:
        print("ERROR: Could not find or split data. Exiting.")
        return

    print("Processing test files...")
    test_data = process_files_to_feature_dict(test_files_dict, concatenate=False, **feature_params)
    print("Processing full training files...")
    full_train_data = process_files_to_feature_dict(train_files_dict, concatenate=True, **feature_params)

    if not test_data or not full_train_data:
        print("ERROR: Feature extraction failed for train/test sets. Exiting.")
        return
    print(f"Data split and processed. Training on {len(full_train_data)}, testing on {len(test_data)} speakers.")

    # --- Experiment (i): Conclusive Effect of Unbalanced Dataset ---
    print("\n(i) Testing conclusive effect of unbalanced dataset")
    imbalance_levels = [0.0, 0.25, 0.50, 0.75]
    unbalanced_results = {}
    speakers_to_reduce = random.sample(list(full_train_data.keys()), k=len(full_train_data) // 2)
    print(f"Will reduce data for speakers: {speakers_to_reduce}")

    for level in imbalance_levels:
        unbalanced_train_data = {}
        for spk, feats in full_train_data.items():
            if spk in speakers_to_reduce and level > 0:
                reduce_to_frames = int(feats.shape[0] * (1 - level))
                if reduce_to_frames > 0:
                    unbalanced_train_data[spk] = feats[:reduce_to_frames, :]
            else:
                unbalanced_train_data[spk] = feats

        gmms = train_gmms(unbalanced_train_data, args.n_mix_unbalanced, verbose=False, **gmm_params)
        eer, fpr, fnr, _ = evaluate_eer_per_clip(gmms, test_data, verbose=False)
        unbalanced_results[f"{int(level*100)}% reduction"] = (eer, fpr, fnr)
        print(f"Imbalance Level {level*100}% -> EER = {eer*100:.2f}%")

    # ... (plotting for imbalance experiment) ...

    # --- Experiment (ii): Effect of Number of Mixture Components & training time ---
    print("\n(ii) Effect of number of mixture components & training time")
    n_components_list = [int(n) for n in args.n_mix_list.split(',')]
    eers_mix, training_times = [], []
    for n_mix in n_components_list:
        start_time = time.time()
        gmms = train_gmms(full_train_data, n_mix, **gmm_params)
        duration = time.time() - start_time
        training_times.append(duration)
        eer, _, _, _ = evaluate_eer_per_clip(gmms, test_data)
        eers_mix.append(eer)
        print(f"n_mix={n_mix}, EER={eer*100:.2f}%, Training Time={duration:.2f}s")

    # ... (plotting for EER and Training time) ...

    # --- Final Evaluation: Confusion Matrix ---
    print("\n(iv) Generating final confusion matrix on test set")
    gmms_final = train_gmms(full_train_data, args.n_mix_unbalanced, **gmm_params)
    if gmms_final:
        evaluate_and_plot_confusion_matrix(gmms_final, test_data)

    print("\nâœ… Experiment completed successfully!")


if __name__ == "__main__":
    parser = argparse.ArgumentParser(description="Advanced GMM-based Speaker Recognition Experiment")

    # General arguments
    parser.add_argument("--data_root", type=str, default="data/data_clips", help="Root directory of speaker data.")
    parser.add_argument("--test_size", type=float, default=0.2, help="Proportion of the dataset for testing.")

    # Feature extraction arguments
    parser.add_argument("--n_mfcc", type=int, default=20, help="Number of MFCCs to extract.")
    parser.add_argument("--n_fft", type=int, default=2048, help="FFT window size for MFCC.")
    parser.add_argument("--hop_length", type=int, default=512, help="Hop length for MFCC.")
    parser.add_argument("--n_mels", type=int, default=128, help="Number of Mel bands for MFCC.")
    parser.add_argument("--no_deltas", action='store_true', help="Disable delta features.")
    parser.add_argument("--no_vad", action='store_true', help="Disable VAD.")

    # GMM training arguments
    parser.add_argument("--n_mix_list", type=str, default="2,4,8,16", help="Comma-separated list of mixture components to test.")
    parser.add_argument("--n_mix_unbalanced", type=int, default=8, help="Mixture components for imbalance experiment.")
    parser.add_argument("--covariance_type", type=str, default="diag", choices=['full', 'tied', 'diag', 'spherical'], help="Covariance type for GMMs.")
    parser.add_argument("--max_iter", type=int, default=200, help="Max EM iterations for GMM training.")
    parser.add_argument("--n_init", type=int, default=3, help="Number of initializations for GMM training.")

    args, unknown = parser.parse_known_args()
    main(args)

```
